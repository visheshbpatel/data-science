```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Colored Scatterplots

```python
iris = pd.read_csv('datasets/iris.csv')
iris.head()
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({
    'Iris-setosa': 0,
    'Iris-versicolor': 1,
    'Iris-virginica': 2
}).astype(int)
```

```python
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='viridis', alpha=0.8)
plt.xlabel('Sepal Lenght')
plt.ylabel('Petal Lenght')
plt.colorbar()
```

## 2. Plot Size

```python
plt.figure(figsize=(6,4))  # plot size

scatter = plt.scatter(
    iris['SepalLengthCm'],
    iris['PetalLengthCm'],
    c=iris['Species'],
    cmap='viridis'
)

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Sepal vs Petal Length')

plt.colorbar(scatter, label='Species') 
plt.show()
```

## 4. horizontal and vertical lines

```python
plt.figure(figsize=(18,18))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

plt.axhline(130, color='red') # horizontal
# 130 strike rate

plt.axhline(140, color='green') # horizontal
# 140 strike rate

plt.axvline(30, color='red')  # vertical

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

## 5. subplots

```python
# A diff way to plot graphs

batter.head()
```

```python
# normal way

plt.figure(figsize=(6,4))

plt.scatter(batter['avg'], batter['strike_rate'])
plt.title('title')
plt.xlabel('Avg')
plt.ylabel('strike rate')
plt.show()
```

```python
# another way

fig, ax = plt.subplots(figsize=(10,8))



ax.scatter(batter['avg'], batter['strike_rate'])
ax.set_title('title')
ax.set_xlabel('Avg')
ax.set_ylabel('strike rate')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))

scatter = ax.scatter(
    batter['avg'],
    batter['strike_rate'],
    c=batter['avg'],   # color based on avg
    cmap='coolwarm',
    alpha=0.7
)

ax.set_title('Batsman Performance Analysis')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')

fig.colorbar(scatter, ax=ax, label='Average')

ax.grid(True)

plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(8,10))

# First plot
ax[0].scatter(batter['avg'], batter['strike_rate'], alpha=0.7)
ax[0].set_title('Average vs Strike Rate')
ax[0].set_xlabel('Average')
ax[0].set_ylabel('Strike Rate')
ax[0].grid(True)

# Second plot
ax[1].scatter(batter['avg'], batter['runs'], alpha=0.7)
ax[1].set_title('Average vs Runs')
ax[1].set_xlabel('Average')
ax[1].set_ylabel('Runs')
ax[1].grid(True)

plt.tight_layout()   # fixes spacing issue
plt.show()
```

```python
# When you have multiple subplots, labels and titles can overlap, spacing becomes messy
# plt.tight_layout() automatically adjusts spacing.
```

```python
# histogram

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

# Scatter plots
ax[0,0].scatter(batter['avg'], batter['strike_rate'])
ax[0,0].set_title('Avg vs Strike Rate')
ax[0,0].set_xlabel('Avg')
ax[0,0].set_ylabel('Strike Rate')

ax[0,1].scatter(batter['avg'], batter['runs'])
ax[0,1].set_title('Avg vs Runs')
ax[0,1].set_xlabel('Avg')
ax[0,1].set_ylabel('Runs')

# Histograms
ax[1,0].hist(batter['avg'], bins=20)
ax[1,0].set_title('Average Distribution')
ax[1,0].set_xlabel('Avg')

ax[1,1].hist(batter['runs'], bins=20)
ax[1,1].set_title('Runs Distribution')
ax[1,1].set_xlabel('Runs')

plt.tight_layout()
plt.show()
```

```python
# subplot grid created using Matplotlib 

fig = plt.figure(figsize=(8,8))

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter['avg'], batter['strike_rate'], color='red')
ax1.set_title('Avg vs Strike Rate')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batter['runs'], bins=20)
ax2.set_title('Runs Distribution')

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batter['avg'], bins=20)
ax3.set_title('Average Distribution')

plt.tight_layout()
plt.show()
```

```python
# This is a subplot grid created using Matplotlib, 
# where multiple visualizations like scatter plots and histograms are displayed in a single figure using add_subplot()
```

## 8. 3D Surface plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
x
```

```python
y
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
xx.shape
```

```python
yy.shape
```

```python
z = xx**2 + yy**2
z
```

```python
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```
