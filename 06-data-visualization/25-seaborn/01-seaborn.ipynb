### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# scatter plot --> axes level function

sns.scatterplot(tips, x='total_bill', y='tip')
```

```python
sns.scatterplot(tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.scatterplot(tips, x='total_bill', y='tip', hue='sex', style='time')
```

```python
sns.scatterplot(tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
# relplot --> figure level --> square shape
sns.relplot(tips, x='total_bill', y='tip', kind='scatter')
```

```python
sns.relplot(tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```

```python
# Seaborn recommends figure-level functions (like sns.relplot(), sns.catplot(), sns.displot()) over axes-level ones 
# (like sns.scatterplot(), sns.barplot()) mainly because they are higher-level, more flexible, and easier for structured analysis.


# Figure-level functions are recommended because they automatically manage figures, support faceting for multi-variable analysis, 
# provide a consistent high-level API, and simplify complex visualizations compared to axes-level functions.
```

```python
# line plot (we prefer line plot when we have date or time on x axis)

# dataset from plotly
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# axes level function
sns.lineplot(temp_df, x='year', y='lifeExp')
```

```python
# using relplot

sns.relplot(temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue --> style

temp_df = gap[gap['country'].isin(['India','Brazil','Germany','China'])]
temp_df
```

```python
sns.relplot(data = temp_df, kind='line', x='year', y='lifeExp', hue='country', style='continent', size='continent')
```

```python
sns.lineplot(data = temp_df, x='year', y='lifeExp', hue='country', style='continent', size='continent')
```

```python
# since relplot is a fig level function , so it controls the fig and keep the legned outside of the graph
```

```python
# facet plot
# A facet plot in Seaborn is a way to visualize subsets of data across multiple panels based on one or more categorical variables 
# It helps you compare patterns across groups side by side

sns.relplot(tips, x='total_bill', y='tip', kind='scatter', col='sex')
```

```python
sns.relplot(tips, x='total_bill', y='tip', kind='scatter', row='sex')
```

```python
sns.relplot(tips, x='total_bill', y='tip', kind='scatter', col='smoker', row='sex')
```

```python
# facet plot --> figure level function --> work with relplot
# it will not work with scatterplot and lineplot
sns.relplot(tips, x='total_bill', y='tip', kind='line', col='smoker', row='sex')
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level --> displot
# axes level --> histplot --> kdeplot --> rugplot
```

```python
# plotting univariate histogram

sns.histplot(tips, x='total_bill')
```

```python
sns.displot(tips, x='total_bill', kind='hist')
```

```python
# Bin parameters

sns.displot(tips, x='total_bill', kind='hist', bins=30)
```

```python
# It’s also possible to visualize the distribution of a categorical variable using the logic of a histogram
# Discrete bins are automatically set for categorical variables

# countplot
sns.displot(tips, x='day', kind='hist')
```

```python
# hue parameter

sns.displot(tips, x='tip', kind='hist', hue='sex')
```

```python
sns.displot(tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.displot(titanic, x='age', kind='hist',hue='sex', element='step')
```

```python
# faceting using col and row  --> this will not work on histplot function

sns.displot(titanic, x='age', kind='hist',col='sex', element='step')
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(tips, x='total_bill')
```

```python
sns.displot(tips, x='total_bill', kind='kde')
```

```python
# hue --> fill
sns.displot(tips, x='total_bill', kind='kde', hue='sex', fill=True)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

sns.kdeplot(tips, x='total_bill')
sns.rugplot(tips, x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(tips, x='total_bill', y='tip')
```

```python
sns.displot(tips, x='total_bill', y='tip', kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian

sns.kdeplot(tips, x='total_bill', y='tip')
```

### 3. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap

# Plot rectangel data as a color-encoded matrix

gap
```

```python
# converting long data into wide data
temp_df = gap.pivot(index='country', columns='year',values='lifeExp')
temp_df
```

```python
# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(temp_df, annot=True, linewidths=0.5, cmap='summer')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.
```

```python
iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
