```python
# Univariate Analysis (One Variable)
# You analyze only one variable at a time.
# To understand the basic properties of that variable.


# Bivariate Analysis (Two Variables)
# You analyze two variables together to see their relationship.
# To find how one variable affects or relates to another.


# Multivariate Analysis (More than Two Variables)
# You analyze more than two variables simultaneously.
# To understand complex relationships between multiple variables.
```

## Pie Chart

<img src="https://media.geeksforgeeks.org/wp-content/uploads/20220920123803/Piechartexample11.png" width=400>

- Univariate/Bivariate Analysis
- Categorical vs Numerical
- Use Case: To find contibution on a standard scale

```python
# simple data
data = [25,45,100,20,49]
plt.pie(data)
```

```python
data = [25,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data, labels=subjects)
plt.show()
```

```python
# dataset

df = pd.read_csv('datasets/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'])
plt.show()
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%.1f%%')
plt.show()
```

```python
# percentage and colors

plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%.1f%%', colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
colors = ['#4E79A7', '#59A14F', '#F28E2B', '#E15759', '#76B7B2', '#EDC948']

plt.pie(
    df['batsman_runs'],
    labels=df['batsman'],
    autopct='%.1f%%',
    colors=colors
)

plt.show()
```

```python
# exploding shadows

plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%.1f%%', explode=[0.1,0,0,0,0,0.3], shadow=True)
plt.show()
```
