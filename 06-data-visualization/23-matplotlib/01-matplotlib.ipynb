```python
# Univariate Analysis (One Variable)
# You analyze only one variable at a time.
# To understand the basic properties of that variable.


# Bivariate Analysis (Two Variables)
# You analyze two variables together to see their relationship.
# To find how one variable affects or relates to another.


# Multivariate Analysis (More than Two Variables)
# You analyze more than two variables simultaneously.
# To understand complex relationships between multiple variables.
```

```python
# from pandas dataframe

batsman = pd.read_csv('datasets/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# plotting multiple plot

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# lable titile

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# color(hex) and line(widht and style) and marker(size)

plt.plot(batsman['index'], batsman['V Kohli'], color='#40FF47')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#40FF47', linestyle='dashed')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='dotted')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#40FF47', linestyle='dashdot')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='solid')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#40FF47', linestyle='dashed', linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='solid', linewidth=1)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#40FF47', linestyle='dashed', linewidth=3, marker='D')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='solid', linewidth=1, marker='.')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#40FF47', linestyle='dashed', linewidth=3, marker='+', markersize=20)
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='solid', linewidth=1, marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend --> location

plt.plot(batsman['index'], batsman['V Kohli'], color='#40FF47', linestyle='dashed', linewidth=3, label='virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='solid', linewidth=1, marker='o', label='rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()   # default loc --> best
```

```python
# legend --> location

plt.plot(batsman['index'], batsman['V Kohli'], color='#40FF47', linestyle='dashed', linewidth=3, label='virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='solid', linewidth=1, marker='o', label='rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='lower right')
```

```python
# grid

plt.plot(batsman['index'], batsman['V Kohli'], color='#40FF47', linestyle='dashed', linewidth=3, label='virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='solid', linewidth=1, marker='o', label='rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.grid()
```

```python
# show

plt.plot(batsman['index'], batsman['V Kohli'], color='#40FF47', linestyle='dashed', linewidth=3, label='virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='blue', linestyle='solid', linewidth=1, marker='o', label='rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparision')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.grid()

plt.show()    # this is the main function the show the graph , will be usefull for normal text editor
```

```python
# plt.scatter simple function

x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,300,50)

print(x)
print()
print(y)
```
