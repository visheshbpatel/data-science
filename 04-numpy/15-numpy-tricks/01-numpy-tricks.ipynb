## 6. np.where

The numpy.where() function returns the indices of elements in an input array where the given condition is satisfied.

https://numpy.org/doc/stable/reference/generated/numpy.where.html

```python
a
```

```python
# find all indices with value greater than 50

np.where(a>50)
```

```python
a[np.where(a>50)]
```

```python
# replace all values > 50 with 0

# syntax: np.where(condition, true, false)

np.where(a>50, 0, a)
```

```python
np.where(a%2==0, 0, a)
```

## 13. np.corrcoef

Return Pearson product-moment correlation coefficients.

https://numpy.org/doc/stable/reference/generated/numpy.corrcoef.html

```python
salary = np.array([2000, 4000, 25000, 35000, 60000])
experience = np.array([1,3,2,4,2])

np.corrcoef(salary, experience)
```


|            | Salary | Experience |
|------------|--------|------------|
| **Salary** | 1.00   |  0.19     |
| **Experience** |  0.19   | 1.00    |



As experience increases, salary tends to increase, but the relationship is not very strong

## 18. Set functions

- np.union1d
- np.intersect1d
- np.setdiff1d
- np.setxor1d
- np.in1d

```python
m = np.array([1,2,3,4,5])
n = np.array([3,4,5,6,7])

np.union1d(m,n)          # all without repetation
```

```python
np.intersect1d(m,n)   # common
```

```python
np.setdiff1d(m,n)    # in m, not in n
```

```python
np.setdiff1d(n,m)    # in n, not in m
```

```python
np.setxor1d(m,n)     # remove common and keep remaning
```
