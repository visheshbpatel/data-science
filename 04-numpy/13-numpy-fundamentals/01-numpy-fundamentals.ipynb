## 1. Numpy Theory

### What is numpy?

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.


At the core of the NumPy package, is the ndarray object. This encapsulates n-dimensional arrays of homogeneous data types

### Numpy Arrays Vs Python Sequences

- NumPy arrays have a fixed size at creation, unlike Python lists (which can grow dynamically). Changing the size of an ndarray will create a new array and delete the original.

- The elements in a NumPy array are all required to be of the same data type, and thus will be the same size in memory.

- NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences.

- A growing plethora of scientific and mathematical Python-based packages are using NumPy arrays; though these typically support Python-sequence input, they convert such input to NumPy arrays prior to processing, and they often output NumPy arrays.

#### NumPy Made in C and Wrapped with Python

NumPy is mostly written in **C (and some C++)** and exposed to us through Python.

Python acts as a wrapper/interface layer, while the heavy computation runs in C.

---

#### 1. What Does "Written in C and Wrapped in Python" Mean?

When you write:

```python
import numpy as np
a = np.array([1, 2, 3])
```

You are calling Python functions.

But internally:

- Python calls C functions  
- The computation runs in C  
- The result is returned back to Python  

So NumPy feels like Python, but runs at C speed.

---

#### 2. Why Is NumPy Written in C?

NumPy is written in C because:

1. C is much faster than pure Python  
2. It allows direct memory management  
3. It stores data in contiguous memory  
4. Mathematical operations can be optimized at low level  

If NumPy were written purely in Python, it would be slow like normal Python loops.

---

#### 3. How Is C Connected to Python?

Python connects to C using:

- CPython C API  
- C extension modules  
- Cython (in some parts)  

Example:

```python
a.sum()
```

Internally:

- Python calls a C function  
- The loop runs in C  
- The result is returned to Python  

This is why NumPy is much faster than Python loops.

---

#### 4. Speed Comparison Example

Pure Python:

```python
sum([1, 2, 3, 4, 5])
```

NumPy:

```python
np.sum(np.array([1, 2, 3, 4, 5]))
```

For large arrays, NumPy is much faster because the loop runs in C instead of Python.

---

#### 5. Important Insight

Most major ML libraries follow this structure:

- NumPy --> C backend  
- TensorFlow --> C++ backend  
- PyTorch --> C++ backend  
- OpenCV --> C++ backend  

Python is used for:

- Clean and readable code  
- Easy API  
- Fast development  

C/C++ is used for:

- Speed  
- Heavy computation  
- Efficient memory handling  

---

#### 6. Memory Difference: NumPy Array vs Python List

##### NumPy Array

- Stored in contiguous memory  
- Fixed data type  
- No object overhead per element  
- Very memory efficient  

##### Python List

- Stores references to objects  
- Each element is a Python object  
- More memory usage  
- Slower for numerical computation  

---

#### Final Summary

NumPy is mainly written in C and wrapped with Python so that:

- It feels like Python  
- But runs at C speed

## 3. Array Attributes

````python
a1 = np.arange(10)
a2 = np.arange(12, dtype=float).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2)

print(a1,'\n')
print(a2,'\n')
print(a3,'\n')
````

````python
# ndim (no. of dimensions)

print(a1.ndim)    #1D
print(a2.ndim)    #2D
print(a3.ndim)    #3D

# .ndim returns the number of axes or dimensions of a NumPy array.
# Important: Dimension means how many directions you need to move to access elements.
````

````python
# shape

print(a1.shape)
print(a2.shape)
print(a3.shape)

# .shape returns a tuple that tells the size of the array in each dimension.
# It tells how many elements exist along each axis.
````

````python
# size

print(a1.size)
print(a2.size)
print(a3.size)

# .size returns the total number of elements in the array.
# It counts all elements across all dimensions.
````

````python
# itemsize

print(a1.itemsize)
print(a2.itemsize)
print(a3.itemsize)


# .itemsize returns the size in bytes of each element in the array.
# Important: It tells memory used by one element, not the whole array.
````

````python
a = np.array([1, 2, 3], dtype=np.int32)
print(a.itemsize)

# int64  -> 8 bytes
# int32  -> 4 bytes
# float64 -> 8 bytes
# float32 -> 4 bytes
````

````python
# dtype

print(a.dtype)
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)

# .dtype tells the data type of elements stored in the NumPy array.
# It tells what type of data each element is stored as internally.
````

````python
a = np.array([1, 2.5, 3])
print(a.dtype)

# Why float
# Because NumPy promotes smaller type to larger compatible type.
# This is called type promotion.
# 8 bits = 1 byte
````

Axis number tells which dimension is being reduced.

In 2D:

Axis 0 collapses rows and keeps columns.

Axis 1 collapses columns and keeps rows.

That is the correct conceptual way to think.

0 means column wise and 1 means row wise

#### Dot Product (Matrix Multiplication)

The dot product is a method of multiplying two matrices following specific rules.

---

#### Rules of Dot Product

##### 1. Inner Dimensions Must Match

If matrix A has shape (m, n)  
And matrix B has shape (n, p)  

Then multiplication is possible.

If the inner dimensions do not match, multiplication is not allowed.

---

##### 2. Shape of Result

If A is (m, n)  
And B is (n, p)  

Then result will have shape (m, p)

Rows come from the first matrix.  
Columns come from the second matrix.

---

##### 3. Order Matters

Matrix multiplication is not commutative.

A × B is not equal to B × A

---

##### 4. It Is Not Element Wise Multiplication

Dot product multiplies rows with columns and then adds the results.

It is different from element wise multiplication using the asterisk operator.

---

#### How Dot Product Is Performed

Suppose:

Matrix A:

```
[[1 2]
 [3 4]]
```

Matrix B:

```
[[5 6]
 [7 8]]
```

To calculate the first element of the result:

Take first row of A and first column of B.

Multiply corresponding elements and add them:

(1 × 5) + (2 × 7)

To calculate the next element:

Take first row of A and second column of B:

(1 × 6) + (2 × 8)

---

#### Memory Trick

Row of first matrix  
Column of second matrix  

Multiply and add.

---

#### Why It Is Important

Dot product is used in:

- Neural networks  
- Linear regression  
- Logistic regression  
- Machine learning models  

Most ML computations rely on matrix multiplication.

## 10. Stacking

````python
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)

print(a4, '\n')
print(a5, '\n')
````

````python
# Horizontal Stacking

np.hstack((a4,a5))

# Important rule:
# Number of rows must be the same.
````

````python
# Vertical Stacking

np.vstack((a4,a5))

# Important rule:
# Number of columns must be the same.
````

```
hstack:
Adds horizontally
Columns increase

vstack:
Adds vertically
Rows increase
```
