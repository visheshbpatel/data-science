#### 4. Why no need to specify data type in Python? How does Python know it?

#### 5. What is Static and Dynamic Typing?

| Typing Type        | Meaning                                                      | Example Language   |
| ------------------ | ------------------------------------------------------------ | ------------------ |
| **Static Typing**  | Data type is checked at compile time. You must declare type. | C, Java            |
| **Dynamic Typing** | Data type is checked at runtime. No need to declare type.    | Python, JavaScript |


- Python = Dynamically Typed



#### 6. What is Static and Dynamic Binding?

- Static Binding (Early Binding)

Binding happens at compile time.

Function calls and types are fixed early.

--

- Dynamic Binding (Late Binding)

Binding happens at run time.

Python uses dynamic binding.

#### 9. What is Compilation?

Compilation is the process of converting high-level code (Python/C/Java) into machine-understandable form.

#### 10. What are compilers and interpreters? Interpreter and compiler languages?

- Compiler

    - Converts whole program at once

    - Faster execution

    - Example languages: C, C++, Java

- Interpreter

    - Executes code line by line

    - Slower but easy to debug

    - Example languages: Python, JavaScript

- Python uses:

    - Interpreter (main)

    - Compiler (internal bytecode compiler)
