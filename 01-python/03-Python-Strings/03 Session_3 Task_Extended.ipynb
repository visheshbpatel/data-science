### 2. Extract username from a given email, Eg if the email is xyz123@gmail.com , then the username should be xyz123

```python
email = input("Enter the Email")

pos = email.index("@")
username = email[0:pos]

print("Username:",username)
```

### 3. Count the Frequency of a perticular character in a provided string, Eg. "Hello how are you" is the string, the frequency of h in this string is 2.

```python
string = input("Enter the String")
char = input("Enter the char for calculating the frequency")
count=0

for i in string:
    if i==char:
        count+=1

print("Count of the",char,"is:",count)
```

### 4. Write a program which can remove a particular character from a string

```python
string = input("Enter the String:")
remove = input("Enter the string you want to remove:")
new_string=""

for i in string:
    if i!=remove:
        new_string = new_string + i

print("New String:",new_string)
```

### 5. Write a program that can check whether a given string is palindrome or not.

```python
word = input("Enter the Word:")

length = len(word)
last = length // 2

palindrome = True

for i in range(0, last):
    if word[i] != word[length - i - 1]:
        palindrome = False
        break

if palindrome == True:
    print("It's a Palindrome")
else:
    print("It's not a Palindrome")
```

```python
# better Code
word = input("Enter the word:")

if word == word[::-1]:
    print("It's a Palindrome")
else:
    print("It's not a palindrome")
```

### 6. Write a program to count the number of words in a string without split()

```python
text = input("Enter the text")
count=1

for i in text:
    if i==" ":
        count+=1

print("The number of words in text is:",count)
```

```python
text = input("Enter the text")
L = []
temp=""
count=1

for i in text:
    if i!=" ":
        temp=temp+i
    else:
        L.append(temp)
        temp=""
        count+=1

L.append(temp)
print(L)
print("No of words:",count)
```

```python
text = input("Enter the text: ")

count = 0
in_word = False

for ch in text:
    if ch!=" " and not in_word:
        count+=1
        in_word=True
    elif ch==" ":
        in_word=False

print("Number of words:",count)
```

### 7. Write a python program to convert a string to title case without using the title()

```python
"Hi how are you".title()
```

```python
text = input("Enter the Text: ")
L=[]

for i in text.split():
    L.append(i[0].upper()+i[1:].lower())

print(" ".join(L))
```

### 8. Write a program that convert an integer to string

```python
str(4)
```

```python
number = int(input("Enter the number"))
num = number
digits = '0123456789'
result=""

while num!=0:
    digit = digits[num%10]
    result = digit+result
    num=num//10

print(result)
print("Data type:",type(result))
```
