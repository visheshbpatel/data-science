# Function

A function is a reusable block of code that performs a specific task.

Instead of writing the same logic repeatedly, we define it once and reuse it whenever required.

> A function:

- Takes input (optional)
- Does some processing
- Gives an output (optional)

> Types of functions

- Built-in functions – already provided
    - print(), len(), sum()

- User-defined functions – written by you

    - def my_function():

##  5. *args and **kwargs 

args = arguments

*args and **kwargs allow us to pass a variable number of arguments to a function.

### 5.1. *args (Non-keyword / Positional arguments)

- *args allows us to pass any number of Non-keyword / positional arguments to a function.
- The arguments are stored as a tuple.

````python
def multiply(*args):
    product = 1
    
    for i in args:
        product = product*i
    return product
    
print(multiply(2,3,4))
print(multiply(2,3,4,5,6,7))
print(multiply(2,3,4,9,8,7,6,5))

````

````python
def multiply(*args):
    product = 1
    print(args)    #--> this is the proof that values are stored in tuple
    print(type(args))

    print(*args)
    
    for i in args:
        product = product*i   
    return product

print(multiply(2,3,4,9,8,7,6,5))


# we can use any word insted of args , like buddy
# print is also implemented it as same way
````

### 5.2 **kwargs (Keyword arguments)

- **kwargs allows us to pass any number of keyword arguments.
- Keyword arguments mean that they contain a key-value pair, like a Python Dictionary.
- The arguments are stored as a dictionary.

````python
def display(**kwargs):

    for (key,value) in kwargs.items():
        print(key,'->',value)

# we can use anyword insted of kwargs
````

````python
display(India='Delhi', Sri_Lanka='Colombo', Nepal='Katmandu', Russia='Moscow')
````

### 5.3 Points to remember while using *args and **kwargs

- order of the arguments matter (normal -> *args -> **kwargs)
- The word 'args' and 'kwargs' are only a convention, we can use any name of our choice



````python
# Example

def demo(a, b, *args, **kwargs):
    print(a, b)
    print(args)
    print(kwargs)

demo(1, 2, 3, 4, x=10, y=20)
````

## 6. How functions are executed in memory?

Python uses **RAM** and a **call stack** to execute functions.

---

##### 1. Function Definition (`def`)

```python
def add(a, b):
    return a + b
```

* Function code is **stored in memory**
* No execution happens at this stage
* Function name stores a **reference** to the function object

---

##### 2. Function Call

```python
add(2, 3)
```

* A **new stack frame** is created
* Memory is allocated for parameters and local variables

---

##### 3. Function Execution

```python
return a + b
```

* Expression is evaluated
* Result is returned to the caller

---

##### 4. Function Ends

* Stack frame is **removed**
* Local variables are **destroyed**
* Control returns to the calling statement

---

##### Call Stack View

```text
add(2, 3)
a = 2
b = 3
→ return 5
```

---

##### Multiple Function Calls Example

```python
def square(x):
    return x * x

def cube(y):
    return square(y) * y

cube(3)
```

* `cube(3)` → stack frame created
* `square(3)` → new stack frame created
* `square` returns → frame destroyed
* `cube` returns → frame destroyed

---

##### Key Points

* Each function call gets **separate memory**
* Local variables exist **only during execution**
* Call stack follows **LIFO** order

---

##### One-line Summary

> Functions execute using a call stack, and memory is freed after exec


## 7. Varibale Scope

#### Variable Scope

Variable scope defines **where a variable can be accessed** in a program.

---

##### Types of Variable Scope

##### 1. Local Scope

* Variables defined **inside a function**
* Accessible **only within that function**

```python
def demo():
    x = 10  # local variable
    print(x)
```

---

##### 2. Global Scope

* Variables defined **outside all functions**
* Accessible **throughout the program**

```python
x = 10  # global variable

def demo():
    print(x)
```

---

##### 3. Global Keyword

* Used to **modify a global variable inside a function**

```python
x = 5

def demo():
    global x
    x = 10
```

---

##### 4. Nonlocal Scope

* Used in **nested functions**
* Refers to variable in the **nearest enclosing function**

```python
def outer():
    x = 10
    def inner():
        nonlocal x
        x = 20
    inner()
```

---

##### Scope Resolution (LEGB Rule)

1. Local
2. Enclosing
3. Global
4. Built-in

---

##### Key Points

* Local variables have **limited lifetime**
* Global variables exist **throughout program execution**
* `global` and `nonlocal` control variable modification

---

##### What `del square` Does

* `del square` removes the **name (reference)** **`square`**
* It does **not immediately delete the function code**
* The function object is deleted **only if no references exist**

---

##### Case 1: Only One Reference

```python
def square(x):
    return x * x

del square
```

* `square` was the **only reference**
* The function object becomes **unreachable**
* Python’s **garbage collector** removes it from memory

---

##### Case 2: Multiple References

```python
def square(x):
    return x * x

x = square
del square
print(x(5))
```

* The name `square` is removed
* `x` still points to the function
* The function object **remains in memory**
* The function can still be called using `x`

---

##### Important Points

* `del` deletes the **reference**, not the object directly
* An object is deleted only when **no references** point to it
* This applies to **functions and all Python objects**

 #### Returning a Nested Function

##### Key Concept

A nested function **cannot be called directly** from the main block, but it **can be called using a reference returned by another function**.

---

##### Example

```python
def f():
    def x(a, b):
        return a + b
    return x

val = f()(3, 4)
print(val)
```

---

##### What Happens Step-by-Step

1. `f()` is called
2. Inside `f`, a nested function `x` is created
3. `return x` returns a **reference** to the function object
4. The name `x` goes out of scope when `f` finishes
5. The **function object remains in memory** because its reference is returned
6. `f()` now evaluates to the function `x`
7. `(3, 4)` calls the returned function

---

##### Important Clarification

* The **name `x` is destroyed**, not the function itself
* The function object is **not destroyed** as long as a reference exists
* Functions are **objects**, and they can be called using any reference

---

##### Equivalent Code (For Understanding)

```python
func = f()
val = func(3, 4)
```

---

> Even though a nested function’s name goes out of scope, the function object remains in memory if its reference is returned, allowing it to be called later.

## 9. Lambda Function

A Lambda Function is a small anonymous Function.

A lambda function can take any number of arguments, but can only have one expression.

<img src="https://intellipaat.com/blog/wp-content/uploads/2025/03/Lambda-Function-in-Python.jpg">


````python
# x -> x^2

sqr = lambda x: x**2
sqr(2)
````

````python
lambda x: x**2
# It return function itself
````

````python
# x,y -> x+y

a = lambda x,y: x+y
a(5,2)
````

#### Difference between Lambda Function and Normal Function

##### Lambda Function

* Has **no name** (anonymous function)
* Written in **one single line**
* Contains **only one expression**
* The expression value is **returned automatically** (no `return` keyword)
* Generally **not reused** for complex logic

##### Normal Function

* Has a **name**
* Can be written in **multiple lines**
* Can contain **multiple statements**
* Uses the `return` keyword to return a value
* Can be **reused** many times

---

##### Important:

* Lambda functions **do return a value**
* They return the **result of the expression automatically**
* They do **not return a function**, they *are* functions

---

#### Why We Use Lambda Functions

Lambda functions are mainly used when we need a **small, simple function for a short period of time**.

They are commonly used with **Higher Order Functions (HOF)** — functions that take another function as an argument.

---

##### Common Use with Higher Order Functions

```python
nums = [1, 2, 3, 4]

result = list(map(lambda x: x * 2, nums))
print(result)
```

Here:

* `map()` is a **Higher Order Function**
* `lambda x: x * 2` is passed as an argument

---

##### When to Use Lambda

* For **short, simple logic**
* When the function is used **only once**
* With `map()`, `filter()`, `reduce()`

---

##### When NOT to Use Lambda

* For **complex logic**
* When readability is important
* When the function needs reuse

---

> Lambda functions are small anonymous functions mainly used with higher-order functions for short and simple operations.


````python
nums = [1, 2, 3, 4]

result = list(map(lambda x: x * 2, nums))
print(result)

# map(function, iterable)
````

````python
# Check if a string has 'a' 

a = lambda s:'a' in s 

print(a('hello'))
print(a('asus'))
````

````python
# odd or even

a = lambda x:'even' if x%2==0 else 'odd'

print(a(5))
print(a(36))
````

## 10. Higher Order Fuction

> A Higher Order Function is a function that does at least one of the following:

- Takes another function as an argument, or

- Returns a function as its result

````python
# Example

def square(x):
    return x**2

#HOF
def transform(f,L):
    output = []
    for i in L:
        output.append(f(i))

    print(output)

L = [1,2,3,4,5]

transform(square,L)
````

````python
# Insted of this we can use Lambda

transform(lambda x:x**2, L)
````

### 10.1 Map

map() is a Higher Order Function that applies a function to each element of an iterable and returns the result.

> map(function, iterable)
    
- function -- what operation to perform

- iterable -- data (list, tuple, etc.)

---

- map() does not modify the original list

- It returns a map object (iterator)

- We usually convert it to list() or tuple()

- map() is just a short and clean version of this loop.

---

- The output of map() is stored in a map object (iterator) and is generated only when we convert it to a list or iterate over it.


````python
# Square the items of a List

print(map(lambda x:x**2, [1,2,3,4,5]))

print(list(map(lambda x:x**2, [1,2,3,4,5])))    # Type conversion into list
````

````python
# odd/even labelling of list items

L = [1,2,3,4,5]

result = list(map(lambda x:'even' if x%2==0 else 'odd', L))
print(result)

for num, lable in zip(L, result):
    print(num,'->',lable)
````

### 10.2 Filter

> filter() is a Higher Order Function that selects elements from an iterable based on a condition.

- filter(function, iterable)

- function -- condition (must return True or False)

- iterable -- data source

---

- filter() does not modify the original iterable

- It returns a filter object (iterator)

- Values are produced only when iterated

````python
# numbers greater than 5

L = [3,4,5,6,7]

print(list(filter(lambda x:x>5,L)))
print()
print(list(map(lambda x:x>5,L)))

````

````python
# fetch fruits starting with 'a'

fruits = ['apple', 'guava', 'cherry']

list(filter(lambda x: x.startswith('a'), fruits))
````

### 10.3 Reduce

reduce( ) is a Higher Order Function that reduces an iterable to a single value by repeatedly applying a function.

````python
# sum of all item

import functools

functools.reduce(lambda x,y:x+y, [1,2,3,4,5])
````

| Step | x  | y | Result |
| ---- | -- | - | ------ |
| 1    | 1  | 2 | 3      |
| 2    | 3  | 3 | 6      |
| 3    | 6  | 4 | 10     |
| 4    | 10 | 5 | 15     |

- Final result -->  15

````python
# find min

functools.reduce(lambda x,y: x if x<y else y, [23,11,45,10,1])
````

| Step | x  | y  | Result (min) |

| 1    | 23 | 11 | 11           |
| 2    | 11 | 45 | 11           |
| 3    | 11 | 10 | 10           |
| 4    | 10 | 1  | 1            |


- Final result -->  1
