## 1. Nested Loops

A nested loop means a loop inside another loop.

- The outer loop runs first.

- For each iteration of the outer loop, the inner loop runs completely.

```python
for i in range(1,5):
    for j in range(1,5):
        print(i,j)
```

### 1.1 Pattern 1

```python
rows = int(input("Enter the number of rows:"))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print("*",end="")
    print()
```

### 1.2 Pattern 2

```python
rows = int(input("Enter the number of rows:"))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end="")
    for k in range(i-1,0,-1):
        print(k,end="")
    print()
```

## 2. Loop Control Statement
- Break
    - break is used to stop the loop immediately, even if the loop condition is still true.
- Continue
    - continue is used to skip the current iteration and move to the next one.
- Pass
    - pass does nothing.
    It is used as a placeholder when a statement is required syntactically but you donâ€™t want to write any code yet.

### 2.1 Break

```python
for i in range(1,10):
    if i ==5:
        break
    print(i)
```

```python
# Finding prime numbers in a given range

lowerRange = int(input("Enter the Lower Range:"))
upperRange = int(input("Enter the Upper Range:"))

for i in range(lowerRange, upperRange+1):
    for j in range(2,i):
        if i%j==0:
          break
    else:
        print(i)
```

### 2.2 Continue

```python
for i in range(1,10):
    if i==5:
        continue
    print(i)
```

### 2.3 Pass

```python
for i in range(1,10):
    pass # will write some code later
    # just a mechanism to avoid error
```
