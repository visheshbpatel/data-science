## 3. methods vs function

> Function

A function is a block of code that exists independently.

- It is not tied to any object

- Defined using def

- Called using its name




> Method

A method is a function that belongs to an object and is defined inside a class.

- It is called using an object

- It operates on the object’s data

- Defined inside a class


````python
L = [1,2,3]

len(L)     #--> function -> because it is outside of the list class

L.append()  #--> method -> beacause it is inside the list class


# len() is a function because it is called independently.
# append() is a method because it is defined inside the list class and called using an object.
````

## 5. Magic Method

A magic method (also called dunder method) is a special method in Python that:

- Starts and ends with double underscores __

- Is automatically called by Python

- Lets your objects work with built-in operations

```python
Example:

__init__ , __str__ , __add__
```



> Why are they called “magic”?

- Because you don’t call them directly most of the time.
- Python calls them behind the scenes.

```python
print(obj)   # calls obj.__str__()
a + b        # calls a.__add__(b)
len(obj)     # calls obj.__len__()
```

### 5.1. Constructor

A constructor is a special method that is automatically called when an object is created.
    
Its main job is to initialize (set up) the object’s data.

    
In Python, a constructor is implemented using the magic method ```__init__```

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
s = Student("Aman", 20)   # constructor is called automatically
```

> What does the constructor do?

- Runs automatically at object creation

- Initializes instance variables

- Prepares the object for use

                      
> Why do we need a constructor?

Without a constructor, every object would start empty.

Constructor ensures:

- Object starts in a valid state

- Required data is provided at creation time

> IMPORANT( ACTUAL MEANING)


***In a constructor, we place initialization and configuration code whose control should not be given to the user.***

***A constructor is used to write configuration-related code that must run automatically when an object is created.***

***For example, when connecting an application to a database (backend), we should not rely on the user to establish the connection; the constructor can handle it automatically.***


## 6. Fraction Class

````python
3/4 * 1/2

# The output is in decimal not in fraction
````

````python
class Fraction:

    def __init__(self,x,y):     #--> parameterized constructor
        self.num = x
        self.den = y

    def __str__(self):
        return '{}/{}'.format(self.num, self.den)

    def __add__(self,other):
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den*other.den
        return '{}/{}'.format(new_num, new_den)

    def __sub__(self,other):
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den*other.den
        return '{}/{}'.format(new_num, new_den)

    def __mul__(self,other):
        new_num = self.num*other.num
        new_den = self.den*other.den
        return '{}/{}'.format(new_num, new_den)

    def __truediv__(self,other):
        new_num = self.num*other.den
        new_den = self.den*other.num
        return '{}/{}'.format(new_num, new_den)

    def covert_to_decimal(self):
        return self.num/self.den
        
````

````python
fr1 = Fraction(3,4)
fr2 = Fraction(1,2)
````
