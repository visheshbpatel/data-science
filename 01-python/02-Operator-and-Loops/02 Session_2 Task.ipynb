### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
ctc = int(input("Enter your CTC(actual): "))

ctc_lakhs = ctc/100000


hra = ctc * 0.10
da = ctc * 0.05
pf = ctc * 0.03


# taxes
if ctc_lakhs < 5:
    tax = 0
elif 5 <= ctc_lakhs < 10:
    tax = ctc * 0.10
elif 10 <= ctc_lakhs <= 20:
    tax = ctc * 0.20
else:
    tax = ctc * 0.30


total_deduction = hra + da + pf + tax
annual_salary = ctc - total_deduction
monthly_salary = annual_salary / 12

print("Your monthly salary after dedctions is:",monthly_salary)
````

````python
# Item 1 Input
item1 = input("Enter the name of item 1:")
cp_item1 = int(input("Enter the cost price of item 1: "))
sp_item1 = int(input("Enter the selling price of item 1: "))
print(" ")


# Item 1 Input
item2 = input("Enter the name of item 2:")
cp_item2 = int(input("Enter the cost price of item 2: "))
sp_item2 = int(input("Enter the selling price of item 2: "))
print(" ")



# Item 1 Individual record
print("Item 1:",item1)
print("Cost Price:",cp_item1)
print("Selling Price:",sp_item1)

if cp_item1 == sp_item1:
    print("There is no profit or loss on item 1:",(item1))
elif cp_item1 < sp_item1:
    print("Profit on item 1:",(item1),"of:",sp_item1 - cp_item1,"!!!")
else:
    print("Loss on item 1:",(item1),"of:",cp_item1 - sp_item1)
print(" ")



# Item 2 Individual record
print("Item 2:",item2)
print("Cost Price:",cp_item2)
print("Selling Price:",sp_item2)

if cp_item2 == sp_item2:
    print("There is no profit or loss on item 2:",(item2))
elif cp_item2 < sp_item2:
    print("Profit on item 2:",(item2),"of:",sp_item2 - cp_item2,"!!!")
else:
    print("Loss on item 2:",(item2),"of:",cp_item2 - sp_item2)
print(" ")


# Item 1 and Item 2 Combined record

total_cp = cp_item1+cp_item2
total_sp = sp_item1+sp_item2

if total_cp == total_sp:
    print("No profit and Loss")

elif total_cp < total_sp:
    print("It's a overall Profit of:",total_sp-total_cp,"!!!")
else:
    print("It's a overall Loss of:",total_cp-total_sp)
````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output: 

```bash
120
```

````python
number = int(input("Enter the number to get factorial: "))
fact=1

if number < 0:
    print("Enter Valid number")
    
elif number==0:
    print("Factorial of 0: 1")
    
else:
    for i in range(1,number+1):
        fact=fact*i

    print("Factorial for",number,"is:",fact)
````

### `Problem 12`:  A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps. 
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps. 

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.** 

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
import math

x, y = 0, 0

while True:
    line = input().strip()
    if line == "!":
        break

    direction, steps = line.split()
    steps = int(steps)

    if direction == "UP":
        y += steps
    elif direction == "DOWN":
        y -= steps
    elif direction == "LEFT":
        x -= steps
    elif direction == "RIGHT":
        x += steps

distance = math.sqrt((x * x) +(y * y))
print(round(distance))
````

### `Problem 13`:Write a program to print whether a given number is a prime number or not

### `Problem 14`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
num_range = int(input("Enter the range to print Armstrong number in it: "))


for i in range(0,num_range+1):
    product=0
    num=i
    while num!=0:
        digit = num%10
        product = product+ (digit**3)
        num=num//10

    if product==i:
        print(i,end=" ")
    
````

### `Problem 15`: :Calculate the angle between the hour hand and minute hand.

Note: There can be two angles between hands; we need to print a minimum of two. Also, we need to print the floor of the final result angle. For example, if the final angle is 10.61, we need to print 10.

Input:<br>
H = 9 , M = 0<br>
Output:<br>
90<br>
Explanation:<br>
The minimum angle between hour and minute
hand when the time is 9 is 90 degress.

````python
import math

H = int(input())
M = int(input())

hour_angle = 30 * H + 0.5 * M
minute_angle = 6 * M

angle = abs(hour_angle - minute_angle)
min_angle = min(angle, 360 - angle)

print(int(min_angle))
````

### `Problem 16`:Given two rectangles, find if the given two rectangles overlap or not. A rectangle is denoted by providing the x and y coordinates of two points: the left top corner and the right bottom corner of the rectangle. Two rectangles sharing a side are considered overlapping. (L1 and R1 are the extreme points of the first rectangle and L2 and R2 are the extreme points of the second rectangle).

Note: It may be assumed that the rectangles are parallel to the coordinate axis.

<img src='https://www.geeksforgeeks.org/wp-content/uploads/rectanglesOverlap.png' width='300' height='200'>

````python
def is_overlapping(L1, R1, L2, R2):
    # Check if one rectangle is to the left of the other
    if R1[0] < L2[0] or R2[0] < L1[0]:
        return False

    # Check if one rectangle is above the other
    if L1[1] < R2[1] or L2[1] < R1[1]:
        return False

    return True


L1 = (0, 10)
R1 = (10, 0)

L2 = (5, 5)
R2 = (15, -5)

print(is_overlapping(L1, R1, L2, R2))
````
