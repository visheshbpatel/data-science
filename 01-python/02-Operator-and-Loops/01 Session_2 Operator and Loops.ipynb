### 1.4 Bitwise Operators

````python
# bitwise and
print(2 & 3)

# 2 = 10
# 3 = 11
#     10 = 2
````

````python
# bitwise or
print(2|3)

# 2 = 10
# 3 = 11
#     11 = 3
````

````python
# bitwise not
print(~3)

# 3 = 00000000 00000000 00000000 00000011

# ~ = 11111111 11111111 11111111 11111100

# ~3 = –4
````

````python
# Bitwise Left Shift (<<)

a = 5        # 5 = 00000101
print(a << 1)
print(a << 2)



# 5 << 1 --> 10 (5 * 2 = 10)

# 5 << 2 --> 20 (5 * 4 = 20)
````

````python
# Bitwise Right Shift (>>)

a = 20       # 20 = 00010100
print(a >> 1)
print(a >> 2)



# 20 >> 1 → 10 (20 / 2 = 10)

# 20 >> 2 → 5 (20 / 4 = 5)
````

### 5.3 The Gessing Game

````python
#Generating a random integer between 1 and 100
import random
jackpot = random.randint(1,100)

guess = int(input("Guess the number: "))
count=1

while guess!=jackpot:
    if guess<jackpot:
        print("Galat!!!, Guess higher")
    else:
        print("Galat!!!, Guess Lower")

    guess = int(input("Guess the number: "))
    count+=1

else:
    print("correct Guess")
    print("Total number of attmepts:",count)
````

````python
#Advance Gaming

import random

secret = random.randint(1, 100)

attempts = 7
count = 0

print("Welcome to number guessing game!")

while count < attempts:

    # Input validation loop
    while True:
        try:
            print(" ")
            guess = int(input("Enter the Number: "))
            break
        except ValueError:
            print("Invalid Input!, Please Enter Number only")

    count += 1

    # Correct guess
    if guess == secret:
        print("Correct Guess!!!")
        print("Attempts:", count)
        break

    # Close guess hint (only if more than 2 attempts left)
    if attempts - count > 1 and guess != secret:
        if abs(secret - guess) < 5:
            print("You Were Close")

    # Higher / Lower hints
    if secret < guess:
        print("Wrong!!!, Guess Lower")
    else:
        print("Wrong!!!, Guess Higher")

    print("Attempts Left:", attempts - count)

# Out of attempts
if count == attempts and guess != secret:
    print("Game Over!, You ran out of Attempts!")
    print("Correct Number is:", secret)
````

### 5.4 For Loop

````python
for i in range(1,11):
    print(i)
````

- How does it generate numbers then?

When you run:

for i in range(1, 11):


Python creates a range iterator.

This iterator has:

- - a small internal counter (pointer)

This counter starts at start (1).

Then:

Loop asks --> iterator returns 1, counter moves to 2

Loop asks --> iterator returns 2, counter moves to 3

Loop asks --> iterator returns 3, counter moves to 4

...

Counter reaches 10

Next would be 11, but stop = 11

-->iterator stops

-->loop ends

So Python only keeps one number in memory:

the current number.

````python
# range(start, stop, step)
for i in range(1,11,2):
    print(i)
````

### 6.2 Sequence sum
1/1! + 2/2! + 3/3! + ....

````python
n = int(input("Enter the n:"))
result = 0
fact = 1

for i in range(1,n+1):
    fact = fact*i
    result = result + (i/fact)

print("Result:",round(result,3))
````
