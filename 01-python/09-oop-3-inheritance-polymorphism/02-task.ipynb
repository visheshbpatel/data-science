### `Problem-1:` Class inheritence

Create a **Bus** child class that inherits from the Vehicle class. The default fare charge of any vehicle is seating capacity * 100. If Vehicle is Bus instance, we need to add an extra 10% on full fare as a maintenance charge. So total fare for bus instance will become the final amount = total fare + 10% of the total fare.

Note: The bus seating capacity is 50. so the final fare amount should be 5500. You need to override the fare() method of a Vehicle class in Bus class.

````python
class Vehicle:

    def __init__(self, capacity):
        self.seating_capacity = capacity

    def fare(self):
        fare_amount = self.seating_capacity*100
        return fare_amount
        
class Bus(Vehicle):

    def fare(self):
        fare_amount = self.seating_capacity*100
        # fare_amount = super().fare()                # can also use super method to access the fare and calculate it
        final_amount = fare_amount + fare_amount*0.1
        return final_amount


the_bus = Bus(50)
the_bus.fare()
````

### `Problem-4:` Write a program that has an abstract class Polygon. Derive two classes Rectangle and Triangle from Polygon and write methods to get the details of their dimensions and hence calculate the area.

````python
from abc import ABC, abstractmethod

class Polygon(ABC):

    @abstractmethod
    def get_details(self):
        pass

    @abstractmethod
    def calculate_area(self):
        pass



class Rectangle(Polygon):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def get_details(self):
        print(f'Length of Rectangle: {self.length}')
        print(f'Width of Rectangle: {self.width}')

    def calculate_area(self):
        print(f'Area of Rectangle: {self.length*self.width}')


class Triangle(Polygon):

    def __init__(self, height, base, hyp):
        self.height = height
        self.base = base
        self.hypotenuse = hyp
        
    def get_details(self):
        print(f'Height of Triangle: {self.height}')
        print(f'Base of Triangle: {self.base}')
        print(f'Hypotenuse of Triangle: {self.hypotenuse}')

    def calculate_area(self):
        print(f'Area of Triangle: {0.5*(self.height*self.base)}')



R1 = Rectangle(4,6)

T1 = Triangle(5,4,3)

R1.get_details()
R1.calculate_area()

print()

T1.get_details()
T1.calculate_area()
````

### `Problem-5:` Write a program with class Bill. The users have the option to pay the bill either by cheque or by cash. Use the inheritance to model this situation.

````python
from abc import ABC, abstractmethod

class Bill(ABC):

    def __init__(self, amount):
        self.amount = amount

    @abstractmethod
    def pay(self):
        pass


class Cheque(Bill):

    def pay(self):
        print(f'Payment of Rs {self.amount} made via cheque')


class Cash(Bill):

    def pay(self):
        print(f'Payment of Rs {self.amount} made via cash')


cheque1 = Cheque(5000)

cash1 = Cash(5000)

cheque1.pay()

cash1.pay()



# ---Important---
# b = Bill(5000)     #---> TypeError: Can't instantiate abstract class Bill without an implementation for abstract method 'pay'
# Any class that has an abstract method cannot be instantiated directly.
# Python prevents you from creating objects of abstract classes.

# by writing @abstractmethod--> That decorator tells Python:
# This method must be overridden in a child class.
````

````python
payments = [Cheque(5000), Cash(5000)]

for p in payments:
    p.pay()
````

````python
# More optimized code

class FlexibleDict(dict):

    def __getitem__(self, key):
        
        
        if key in self:
            return super().__getitem__(key)

        try:
            str_key = str(key)
            if str_key in self:
                return super().__getitem__(str_key)

        except:
            pass


        try:
            int_key = int(key)
            if int_key in self:
                return super().__getitem__(int_key)

        except (ValueError, TypeError):
            pass


        # if nothing happens
        raise KeyError(key)



        

fd = FlexibleDict()
fd['a'] = 100
print(fd['a']) # Like regular dict

fd[5] = 500
print(fd[5]) # Like regular dict

fd[1] = 100
print(fd['1']) # actual Key is int but still trying to access through str key.

fd['1'] = 100
print(fd[1])


fd['2'] = 200
print(fd[2])


fd['3'] = 300
# print(fd['30'])   # ---> KeyError: '30'
# this gives error and our code is working correctly
````
