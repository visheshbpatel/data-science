## 1. Aggregation(Has-A Relationship)

Aggregation is a Has-A relationship where:

 - One object uses or refers to another object

 - Both objects can exist independently

 - The relationship is loose


> One class has a reference to another class

***In aggregation, two classes have a Has-A relationship, and while creating an object of the main class, we pass the object of another class to it.***

```python
add1 = Address("Indore", "123456", "MP")   # created independently

cust = Customer("vbp", "male", add1)      # passed to Customer
```

````python
# Example

class Customer:

    def __init__(self, name, gender, address):
        self.name = name
        self.gender = gender
        self.address = address

    def print_address(self):
        print(self.address.get_city(), self.address.pin, self.address.state)


    def edit_profile(self, new_name, new_city, new_pin, new_state):
        self.name = new_name
        self.address.edit_address(new_city, new_pin, new_state)


class Address:

    def __init__(self, city, pin, state):
        self.__city = city                 # This private attributes cant' be accssed by another class
        self.pin = pin
        self.state = state


    def get_city(self):                    # Created getter method to access private city attribute for another class
        return self.__city                 # Encapsulation: We hide data and expose it through methods.


    def edit_address(self, new_city, new_pin, new_state):
        self.__city = new_city
        self.pin = new_pin
        self.state = new_state
        

add1 = Address('Indore', '123456', 'Madhya Pradesh')

cust = Customer('vbp', 'male', add1)         # Both add1 and cust.address point to the same object.

cust.print_address()

cust.edit_profile('buddy', 'Bern',  '111111', 'Switzerland')

cust.print_address()
# method example
# What about private attribute
````

## 2. Inheritance

It represents an IS-A relationship.

Example:

- Avoid duplication

- Logical hierarchy

### 2.4 Super Keyword

***super( ) is used to call methods of the parent class inside the child class.***

> It helps:

- Reuse parent logic

- Avoid rewriting code

- Maintain inheritance hierarchy

````python
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a phone")


class SmartPhone(Phone):
    def buy(self):
        print("Buying a smartphone")
        # syntax to call parents's buy method
        super().buy()


s = SmartPhone(20000, "Apple", 13)
s.buy()
````

````python
# super -> constructor
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera


class SmartPhone(Phone):

    def __init__(self, price, brand, camera, os, ram):
        print("Inside smartphone constructor 1")
        super().__init__(price,  brand, camera)        # triggerd parent's constructor from child's constructor
        self.os = os
        self.ram = ram
        print("Inside smartphone constructor 2")



s = SmartPhone(20000, "Apple", 13, 'Android', 12)
print(s.os)
print(s.brand)

# s.super().buy()      #---> AttributeError: 'SmartPhone' object has no attribute 'super'
# super().buy()      #---> AttributeError: 'SmartPhone' object has no attribute 'super'
# this will not work because super() is used in only inside class
````

````python
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera
````

    
````python
    def buy_me(self):
        print("Buying a phone 1")

    def buy(self):
        print("Buying a phone 2")


class SmartPhone(Phone):
    def buy(self):
        print("Buying a smartphone")
        
        super().buy_me()
        self.buy_me()

        # syntax to call parents's buy method
        super().buy()
        # this super call's the parent class buy not the child
        # this super hepls when we have methods of same name and want to call the method of parent class in child class
        


s = SmartPhone(20000, "Apple", 13)
s.buy()
````

### 2.6 Inheritance Example

````python
class Parent:

    def __init__(self,num):
      self.__num=num

    def get_num(self):
      return self.__num

class Child(Parent):
  
    def __init__(self,num,val):
      super().__init__(num)
      self.__val=val

    def get_val(self):
      return self.__val
      
son=Child(100,200)
print(son.get_num())
print(son.get_val())
````

#### 1. Single Inheritance

One child class inherits from one parent class.

````python

class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    pass

SmartPhone(1000,"Apple","13px").buy()
````

#### 2. Multilevel Inheritance

A class inherits from a class which itself inherits from another class.

````python
class Product:
    
    def review(self):
        print("Product customer review")

class Phone(Product):
    
    def __init__(self, price, brand, camera):
        print("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera


    def buy(self):
        print("Buying a phone")


class SmartPhone(Phone):
    pass


s = SmartPhone(20000, "Apple", 12)

s.buy()
s.review()
````

#### 3. Hierarchical Inheritance

Multiple child classes inherit from the same parent class.

````python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class SmartPhone(Phone):
    pass

class FeaturePhone(Phone):
    pass

SmartPhone(1000,"Apple","13px").buy()
FeaturePhone(10,"Lava","1px").buy()
````

#### Multiple Inheritance

One child class inherits from more than one parent class.

````python
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")


class Product:
    def review(self):
        print ("Customer review")


class SmartPhone(Phone, Product):
    pass

s=SmartPhone(20000, "Apple", 12)

s.buy()
s.review()
````

#### Hybrid Inheritance

Combination of two or more types of inheritance.

#### The Diamond Problem

````python
# https://stackoverflow.com/questions/56361048/what-is-the-diamond-problem-in-python-and-why-its-not-appear-in-python2
class Phone:
    def __init__(self, price, brand, camera):
        print ("Inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print ("Buying a phone")

class Product:
    def buy(self):
        print ("Product buy method")

# Method resolution order(MRO)
class SmartPhone(Phone,Product):
    pass

s=SmartPhone(20000, "Apple", 12)

s.buy()
````

````python
class A:

    def m1(self):
        return 20

class B(A):

    def m1(self):
        return 30

    def m2(self):
        return 40

class C(B):
  
    def m2(self):
        return 20
obj1=A()
obj2=B()
obj3=C()
print(obj1.m1() + obj3.m1()+ obj3.m2())
````

````python
class A:

    def m1(self):
        return 20

class B(A):

    def m1(self):
        val=super().m1()+30
        return val

class C(B):
  
    def m1(self):
        val=self.m1()+20     # --> RecursionError: maximum recursion depth exceeded
        return val
obj=C()
# print(obj.m1())
````


## 3. Polymorphism

Polymorphism means "many forms".

> In OOP, it means:

- The same method name behaves differently for different objects.

### Method Overriding

````python
class A:
    def display(self):
        print("This is class A")

class B(A):
    def display(self):   # Overriding the parent method
        print("This is class B")

obj = B()
obj.display()
````

Class B inherits from A

Both classes have a method named display

Python first looks inside class B

It finds display in B

So it calls B's version

### Method overloading

````python
class Shape:

    def area(self, radius):       # area_circle
        return 3.14*radius*radius

    def area(self,l,b):           # area_rectangle
        return l*b

s = Shape()

# s.area(2)       # ---> TypeError: Shape.area() missing 1 required positional argument: 'b'

s.area(3,4)
````

**In python Method overloading is not allowed** 

````python
# One of the way to  implement method overloading

class Shape:

    def area(self, a,b=0):
        if b==0:
            return 3.14*a*a
        else:
            return a*b


s = Shape()

print(s.area(2))      
print(s.area(3,4))
````

## 4. Abstraction

> What is Abstraction

Abstraction means hiding internal implementation details and showing only essential features to the user.

- User only knows what to do, not how it is done.

---
    

> What is an Abstract Class

An abstract class is a class that cannot be instantiated and is used as a blueprint for other classes.

Important Points:

- Cannot create object of abstract class

- Can have normal methods

- Can have abstract methods

---

> What is an Abstract Method

An abstract method is a method declared in an abstract class but does not contain implementation.

Example:

```python
@abstractmethod
def security(self):
    pass
```

This means:

Every child class MUST implement security()

Otherwise → error.

````python
from abc import ABC, abstractmethod

# Abstract class
class BankApp(ABC):

    def database(self):
        print("Connected to database")

    @abstractmethod
    def security(self):
        pass
````

````python
class MobileApp(BankApp):

    def mobile_login(self):
        print("login into mobile")
````

````python
# mob = MobileApp()

# Error ---> TypeError: Can't instantiate abstract class MobileApp without an implementation for abstract method 'security'
````

````python
# correct program

class MobileApp(BankApp):

    def mobile_login(self):
        print("login into mobile")

    def security(self):
        print("Mobile security")
````

````python
mob = MobileApp()
````

````python
mob.database()
mob.mobile_login()
mob.security()
````
