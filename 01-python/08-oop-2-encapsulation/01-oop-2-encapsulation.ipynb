## 1. Revision of OOP-1

````python
class Point:

    # 1. A user can create and view 2D coordinates
    def __init__(self,x,y):
        self.x_cod = x
        self.y_cod = y

    def __str__(self):
        return '<{},{}>'.format(self.x_cod, self.y_cod)

    
    # 2. A user can find out the distance between 2 coordinates
    def euclidean_distance(self, other):
        return ((self.x_cod - other.x_cod)**2 + (self.y_cod - other.y_cod)**2)**0.5


    # 3. A user can find the distance of a coordinate from origin
    def distance_from_origin(self):
        # return (self.x_cod**2 + self.y_cod**2)**0.5     #--> Another way to write
        return self.euclidean_distance(Point(0,0))



class Line:

    def __init__(self, A, B, C):
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return '{}x + {}y + {} = 0'. format(self.A, self.B, self.C)


    # A user can check if a point liers on a given line
    # Ax+By+C=0
    # (1,1) x+y-2=0 --> (1)+(1)-2=0 , then 0=0 RHS==LHS , the condition of the line is satisfied
    def point_on_line(line, point):     #--> self=line
        if line.A*point.x_cod + line.B*point.y_cod + line.C == 0:
            return 'lies on the line'
        else: 
            return 'Does not lie on the line'


    # A user can find the distance between a given 2D point and a given line
    def shortest_distance(line, point):
        return abs(line.A*point.x_cod + line.B*point.y_cod + line.C) // (line.A**2 + line.B**2)**0.5



    # Check whether two straight lines intersect (TASK)
    def did_line_intersect(line1, line2):
        if line1.A*line2.B - line2.A*line1.B != 0:
            return 'Line Intersects'
        else:
            return 'Line did not Intersects'
````

- Reference variables hold references (addresses) to objects, not the objects themselves.
- we can create objects without reference variable as well
- An object can have multiple reference variabels
- Assigning a new reference variable to an existing object does not create a new object

````python
# Multiple ref
print(id(p))
print(id(q))

# This shows that both p and q points to the same memory address
````

## 4. Pass by reference

````python
class Person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender

# outside the class --> function
def greet(person):
    print('Hi my name is',person.name,'and I am a', person.gender)

p = Person('vbp','male')
greet(p)
````

```python
p ───────┐
         ├──► Person object (name='vbp', gender='male')
person ──┘
```
- so now the p and person(inside greet) both points to class Person

````python
class Person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender

# outside the class --> function
def greet(person):
    print('Hi my name is',person.name,'and I am a', person.gender)
    p1 = Person('buddy','female')     #--> p1 is a local reference variable. This object has no connection to p or person
    return p1     #--> The reference stored in p1 is returned

p = Person('vbp','male')
x = greet(p)     #---> That reference is assigned to x (return p1)
print(x.name)
print(x.gender)
````

```python
p  ───► Person('vbp','male')
x  ───► Person('buddy','female')
```
- Two different objects in memory.

> Passing an object to a function passes its reference, but creating and returning a new object inside the function creates a separate object in memory.

````python
class Person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender

# outside the class --> function
def greet(person):
    print(id(person))
    print('Hi my name is',person.name,'and I am a', person.gender)

p = Person('vbp','male')
print(id(p))
greet(p)


# The id of p and person both object's id are same , so we did't send the object we sent the reference of the object or address of object 
# Same id() inside and outside the function proves that Python passes the reference to the object, not a copy of the object.
````

## 5. Object ki Mutability

````python
class Person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender

# outside the class --> function
def greet(person):
    person.name = 'buddy'
    person.gender='female'
    return person
    
p = Person('vbp','male')
print(id(p))
p1 = greet(p)
print(id(p))
print(id(p1))

print()
print(p.name)
print(p.gender)
````

> A mutable object is one whose internal state can be changed without creating a new object.

- User-defined class objects (like Person) are mutable by default.

- The id is unchanged

- This proves the object was mutated, not replaced


```python
p  ─────┐
        ├──► Person(name='buddy', gender='female')
p1 ─────┘
```



- If the object were immutable, changing name would require creating a new object

- Since the id stays the same, the object itself was modified

> Person objects are mutable because their attributes can be changed without changing their identity (id).

> Encapsulation

Encapsulation is an Object-Oriented Programming (OOP) concept in which data (variables) and methods (functions) that operate on that data are wrapped together inside a class, and direct access to the data is restricted.

- Encapsulation = Data Hiding + Controlled Access

- Encapsulation helps protect the internal state of an object and ensures that data is accessed or modified only in safe and controlled ways.

> Why Encapsulation is Needed:

- Prevents accidental modification of data

- Improves data security

- Maintains data integrity

- Makes code easier to maintain and debug

---
    
> Getter Methods

A getter is a method that is used to access (read) the value of a private instance variable.

Purpose of Getter: 

- Provides controlled read access to data

- Prevents direct access to private variables

Conceptually:

“You can see the value, but you cannot change it directly.”

---

> Setter Methods

A setter is a method that is used to modify (update) the value of a private instance variable with validation rules.

Purpose of Setter: 

- Controls how data is modified

- Prevents invalid values from being assigned

Conceptually:

“You can change the value, but only if it follows my rules.”

---

> Why Getters and Setters are Important

- Prevent direct access to private variables

- Allow validation before updating data

- Keep the object’s state consistent

- Improve code safety and reliability

***`Instance & Static`***

> Instance Variable
- An instance variable is a variable that stores data specific to an object and is created separately for each object of a class.


> Static Variable
- A static variable (also called a class variable) is a variable that belongs to the class, not to individual objects.

- It stores data that is shared by all objects of the class.

````python
# After Implementing Static variable

class Atm:

    __counter = 1     # static (class) variable.

    # constructor (special function) -> superpower -> automatically execute's when object is creatred
    def __init__(self):
        self.pin = ''
        self.__balance = 0      
        self.cid = Atm.__counter          
        Atm.__counter+=1
        # self.menu()

    #utility functions
    @staticmethod
    # this is a staticmethod
    def get_counter():            # A method without self
        return Atm.__counter

    def get_balance(self):
        return self.__balance

    def set_balance(self, new_value):
        if type(new_value)==int:
            self.__balance = new_value

        else:
            print("Beta bahto maaarenge")

    def menu(self):
        user_input = input("""
        Hi how can I help you?
        1. Press 1 to create pin
        2. press 2 to change pin
        3. Press 3 to check balance
        4. Press 4 to withdraw
        5. Anything else to exit
        """)

        if user_input == '1':
            # create pin
            self.create_pin()

        elif user_input == '2':
            self.change_pin()

        elif user_input == '3':
            self.check_balance()

        elif user_input == '4':
            self.withdraw()

        else:
            pass

    def create_pin(self):
        user_pin = input("Enter your pin: ")
        self.pin = user_pin

        user_balance = int(input("Enter balance: "))
        self.__balance = user_balance


        print('pin created successfully')
        self.menu()

    def change_pin(self):
        old_pin = input("Enter your old pin: ")

        if old_pin == self.pin:
            new_pin = input("Enter new pin: ")
            self.pin = new_pin
            print("Pin changed successfully")
            self.menu()
            
        else:
            print("Nahi karne de sakta re baba")
            self.menu()


    def check_balance(self):
        user_pin = input("Enter your Pin: ")
        if user_pin == self.pin:
            print("Your balance is: ", self.__balance)
        else:
            print("chal nikal yahan se")


    def withdraw(self):
        user_pin = input("Enter the Pin: ")
        if user_pin == self.pin:
            amount = int(input("Enter the amount: "))
            if  amount<=self.__balance:
                self.__balance = self.__balance - amount
                print("withdrwal successfully, balance is: ", self.__balance)
            else:
                print("abe garib")
        else:
            print("Sale chor")
            
````

***`Instance Vs  Static`***

| Instance Variable                           | Static (Class) Variable                 |
| ------------------------------------------- | --------------------------------------- |
| Belongs to an object (instance)             | Belongs to the class                    |
| Created when an object is created           | Created when the class is loaded        |
| Separate copy for each object               | Single shared copy for all objects      |
| Stored inside each object                   | Stored inside the class                 |
| Not shared between objects                  | Shared by all objects                   |
| Accessed using object reference (`obj.var`) | Accessed using class name (`Class.var`) |
| Change affects only that object             | Change affects all objects              |
| Used for object-specific data               | Used for common/shared data             |
| Memory allocated per object                 | Memory allocated once                   |
| Initialized using `self`                    | Defined directly inside class           |
| Exists as long as object exists             | Exists as long as class exists          |


***`Utility Functions`***

Utility functions are functions that:

- Perform a general-purpose task

- Do not depend on object state

- Are often reused in many places


*Utility functions do some work but do not need object data (self).*

***`Static Method`***

A static method is a method inside a class that:

- Does not receive self

- Does not receive cls

- Belongs logically to the class

- Is used via the class name


*Static methods are utility functions placed inside a class for better organization.*



> How Python identifies static methods

Using the decorator: @staticmethod

*This tells Python: “Do not pass self automatically.”*
