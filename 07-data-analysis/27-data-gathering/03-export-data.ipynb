# Pandas Export

```python
import pandas as pd
```

## 1. to_csv

```python
df = pd.read_csv("datasets/deliveries.csv")
df.head()
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df
```

```python
temp_df.to_csv('batsman-runs.csv', index=False)
```

```python
df.pivot_table(index='batsman', columns='bowling_team', values='batsman_runs', aggfunc='sum')
```

```python
df.pivot_table(index='batsman', columns='bowling_team', values='batsman_runs', aggfunc='sum').to_csv('batsman_vs_team.csv',index=True)
```

## 2. to_excel

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df
```

```python
temp_df.to_excel('batsman-runs.xlsx')
```

```python
temp_df.to_excel("batsman-runs.xlsx", sheet_name='batsman_runs')
```

```python
temp_df2 = df.pivot_table(index='batsman', columns='bowling_team', values='batsman_runs', aggfunc='sum')
```

```python
with pd.ExcelWriter('output.xlsx') as writer:
    temp_df.to_excel(writer, sheet_name = "sheet_name1")
    temp_df2.to_excel(writer, sheet_name = "sheet_name2")
```
