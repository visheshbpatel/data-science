## `Q-8` Batsman record season wise

Make a function which takes a input `batsman_name` and it returns a dataframe.
Columns of the data frame are - `['Season','Innings', 'TotalRuns', 'Avg', 'HighestScore','StrikeRate']`.
* In result make `Season` column as index.

* Avg - total_runs/ no of time got out. - player_out column will help.
* StrikeRate -(total_runs/ balls faced) * 100- wides are not included in batsman ball faced counts. No balls are included. -> Extra_type column will help
* Batsman Can score runs on No Balls.
* Batsman can get out on No Ball or Wides. And even while being on non-striker. Keep these things in mind before masking.

````python
deliveries.head()
````

````python
ipl.head()
````

````python
def batsman_record(batsman_name):
    
    # merge datasets
    df = deliveries.merge(ipl, on='ID')
    
    # filter batsman
    df = df[df['batter'] == batsman_name].copy()
    

    
    # 1. Total Runs (season-wise)
    total_runs = df.groupby('Season')['batsman_run'].sum()
    

    
    # 2. Innings (matches played)
    innings = df.groupby('Season')['ID'].nunique()
    

    
    # 3. Balls faced (exclude wides only)
    valid_balls = df[df['extra_type'] != 'wides']
    balls = valid_balls.groupby('Season').size()
    

    
    # 4. Number of times OUT
    outs = df[df['player_out'] == batsman_name].groupby('Season').size()
    

    
    # 5. Highest Score (per match, then max per season)
    highest_score = (df.groupby(['Season','ID'])['batsman_run'].sum().groupby('Season').max())
    

    
    # Combine all
    result = pd.concat(
        [innings, total_runs, outs, highest_score, balls],
        axis=1
    )
    
    result.columns = ['Innings','TotalRuns','Outs','HighestScore','Balls']
    

    
    # 6. Average
    result['Avg'] = result['TotalRuns'] / result['Outs']


    
    # handle division by zero (not out season)
    result['Avg'] = result['Avg'].fillna(result['TotalRuns'])
    

    
    # 7. Strike Rate
    result['StrikeRate'] = (result['TotalRuns'] / result['Balls']) * 100


    
    # Final formatting
    result = result[['Innings','TotalRuns','Avg','HighestScore','StrikeRate']]
    
    result.sort_values(by='Season')
    
    return result

    

batsman_record("V Kohli")
````

## `Q-9` Using both dataset, make a dataframe as described below

Data Frame columns-> `['PlayerOfThematch', 'BattingFigure', 'BowlingFigure']`

* BattingFigure->`<runs>/<balls>`
* BowlingFigure->`<wicket>/<runs-conceded>`

DataFrame should have one record for each match.

Say 'V Kohli' got POM award then in dataset include his batting figure of that match. Say he scored 112runs in 76 balls. And he hasn't bowled so Bowling Figure will be NaN
```
PlayerOfThematch BattingFigure BowlingFigure
V Kohli          112/76         nan  

```


````python
def pom_performance():
    
    # merge datasets
    df = deliveries.merge(ipl, on='ID')
    
    # -----------------------------
    # 1. Batting stats per match-player
    bat_df = df.groupby(['ID','batter']).agg({
        'batsman_run': 'sum'
    }).rename(columns={'batsman_run': 'runs'})
    
    # balls faced (exclude wides)
    balls = df[df['extra_type'] != 'wides'].groupby(['ID','batter']).size().rename('balls')
    
    bat_df = bat_df.join(balls)
    
    # create batting figure
    bat_df['BattingFigure'] = bat_df['runs'].astype(int).astype(str) + '/' + bat_df['balls'].astype(str)
    
    # -----------------------------
    # 2. Bowling stats per match-player
    bowl_df = df.groupby(['ID','bowler']).agg({
        'isWicketDelivery': 'sum',
        'total_run': 'sum'
    }).rename(columns={
        'isWicketDelivery': 'wickets',
        'total_run': 'runs_conceded'
    })
    
    # create bowling figure
    bowl_df['BowlingFigure'] = bowl_df['wickets'].astype(int).astype(str) + '/' + bowl_df['runs_conceded'].astype(int).astype(str)
    
    # -----------------------------
    # 3. Get Player of the Match
    pom = ipl[['ID','Player_of_Match']].rename(columns={'Player_of_Match':'PlayerOfTheMatch'})
    
    # -----------------------------
    # 4. Merge everything
    
    # batting merge
    result = pom.merge(
        bat_df[['BattingFigure']],
        left_on=['ID','PlayerOfTheMatch'],
        right_on=['ID','batter'],
        how='left'
    )
    
    # bowling merge
    result = result.merge(
        bowl_df[['BowlingFigure']],
        left_on=['ID','PlayerOfTheMatch'],
        right_on=['ID','bowler'],
        how='left'
    )
    
    # -----------------------------
    # final output
    return result[['PlayerOfTheMatch','BattingFigure','BowlingFigure']]


pom_performance()
````

### `Q-11:` Find out the maximum and minimum sepal width and petal width of Setosa and Versicolor. To do this:
- First create a dataset with merging the required datasets
- After that, use `groupby` to create groups based on the "Species" column.
- Then find out which are asked in this question.


The output should be like this:
```bash
Minimum Sepal width of Setosa is 2.3
Maximum Sepal width of Setosa is 4.4

**************************************************

Minimum Sepal width of Versicolor is 2.0
Maximum Sepal width of Versicolor is 3.4

**************************************************
```

````python
# Step 1: Combine only required species
species_df = pd.concat([setosa, versicolor], ignore_index=True)


# Step 2: Merge all required datasets
df = sepal_all.merge(petal_all, on='Id')
df = df.merge(species_df, on='Id')


# Step 3: Drop unwanted columns
df.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y', 'Unnamed: 0'], inplace=True)


# Step 4: Group by Species
grouped = df.groupby('Species')


# Step 5: Calculate min & max
result = grouped[['SepalWidthCm', 'PetalWidthCm']].agg(['min', 'max'])


# Step 6: Print in required format
for species in ['Iris-setosa', 'Iris-versicolor']:
    print(f"Minimum Sepal width of {species.split('-')[1].capitalize()} is {result.loc[species, ('SepalWidthCm', 'min')]}")
    print(f"Maximum Sepal width of {species.split('-')[1].capitalize()} is {result.loc[species, ('SepalWidthCm', 'max')]}")
    
    print()
    print("*" * 50)
    print()
````
