### 3. Your are given a file items.csv which has item_id and item_name. Find out most sold items in each quarter.


````python

most_sold_q1 = q1.merge(items, how='inner', on='item_id').groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(1)
most_sold_q2 = q2.merge(items, how='inner', on='item_id').groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(1)
most_sold_q3 = q3.merge(items, how='inner', on='item_id').groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(1)

print('q1',most_sold_q1, '\n')
print('q2',most_sold_q2, '\n')
print('q3',most_sold_q3)
````

````python
# 3. Your are given a file items.csv which has item_id and item_name. Find out most sold items in each quarter.

def most_sold(df):
    return (df.merge(items, on='item_id')
              .groupby('item_name')['quantity']
              .sum()
              .sort_values(ascending=False)
              .head(1))

print('q1 most sold: ')
print(most_sold(q1), '\n')
print('q2 most sold: ')
print(most_sold(q2), '\n')
print('q3 most sold: ')
print(most_sold(q3))
````

````python
def most_revenue(df):
    df = df.copy() 
    df['revenue'] = df['item_price'] * df['quantity']
    df_revenue = df.merge(items, on='item_id').groupby('item_name')['revenue'].sum().sort_values(ascending=False)
    df_most_revenue = df_revenue.head(1)
    return df_most_revenue


print('q1 most revenue: ')
print(most_revenue(q1), '\n')

print('q2 most revenue: ')
print(most_revenue(q2), '\n')

print('q3 most revenue: ')
print(most_revenue(q3))
````

### 5. Find out avg order price of each quarter.


````python
q1.head(3)
````

````python
# 5. Find out avg order price of each quarter.


def avg_order_price(df):
    df = df.copy()
    df['revenue'] = df['item_price'] * df['quantity']
    df_avg_price = df['revenue'].sum() / df['order_id'].nunique()
    return df_avg_price


print('q1 avg order price: ', end='')
print(avg_order_price(q1), '\n')

print('q2 avg order price: ', end='')
print(avg_order_price(q2), '\n')
````

## `Q-6` From the IPL wala dataset you have to find the Purple cap holder each season.

*Note: Bowler with most no wickets in a season gets purple cap. If more than one bowler have same no of wickets in the season, one with least ecomnomy among them is purple cap holder.*

Bowler's Economy = runs-conceded per six balls

````python
deliveries = pd.read_csv('datasets/data/ipl_deliveries.csv')
deliveries
````

````python
ipl = pd.read_csv('datasets/data/IPL_Matches_2008_2022.csv')
ipl
````

````python
new_ipl = deliveries.merge(ipl, on='ID')
new_ipl
````

````python
purple = new_ipl.groupby(['Season','bowler']).agg({
    'total_run': 'sum',
    'isWicketDelivery': 'sum',
    'batsman_run': 'count'
}).reset_index()


purple = purple.rename(columns={'batsman_run': 'balls'})


purple['economy'] = (purple['total_run'] / purple['balls']) * 6


purple = purple.sort_values(
    ['Season','isWicketDelivery','economy'],
    ascending=[True, False, True]
)


purple_cap = purple.groupby('Season').head(1)

purple_cap
````
