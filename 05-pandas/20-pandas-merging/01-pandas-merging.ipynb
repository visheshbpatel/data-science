### 1. inner join

````python
students.head(3)
````

````python
students.shape
````

````python
regs.head(3)
````

````python
regs.shape
````

````python
# inner join

# first is left table and second is right table
# first       #second
students.merge(regs, how='inner', on='student_id')
````

```
students is the left table
regs is the right table
on='student_id' means matching rows where student_id is same
how='inner' means:
    only common student_id values will be kept
    non-matching rows are removed

# Inner join: keeps only matching student_id in both tables

```

### 2. left join

````python
# left join

courses.merge(regs, how='left', on='course_id')
````

```
courses is the left table
regs is the right table
Matching happens on course_id
how='left' means:
    All rows from courses are kept
    Matching rows from regs are added
    If no match is found, values from regs become NaN

```

### 3. right join

````python
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','vbp','buddy'],
    'partner':[25,28,27]
})

temp_df
````

````python
students = pd.concat([students,temp_df], ignore_index=True)
students
````

````python
students.tail(3)
# this 3 students have not enrolled in any course
````

````python
# right join

students.merge(regs, how='right', on='student_id')
````

```
students is the left table
regs is the right table
Matching is done using student_id
how='right' means:
    All rows from regs are kept
    Matching rows from students are added
    If no match is found, values from students become NaN
```

### 4. outer join

````python
# outer join

students.merge(regs, how='outer', on='student_id')
````

````python
students.merge(regs, how='outer', on='student_id').tail(10)
````

```
Takes all student_id from both tables
Matches rows where possible
Where no match exists, fills with NaN
```

### crux

```
1. Inner Join
Keeps only the rows that are present in both tables

2. Left Join
Keeps all rows from the left table and matching rows from the right table
If no match, right side becomes NaN

3. Right Join
Keeps all rows from the right table and matching rows from the left table
If no match, left side becomes NaN

4. Outer Join
Keeps all rows from both tables
Non-matching values are filled with NaN
```
---

| Join Type | Rows Kept           | Non-matching Handling               | Depends on Order | Result Size   | Simple Meaning    |
| --------- | ------------------- | ----------------------------------- | ---------------- | ------------- | ----------------- |
| Inner     | Only common rows    | Removed                             | No               | Smallest      | Only overlap      |
| Left      | All rows from left  | Right side becomes `NaN`            | Yes              | ≥ left table  | Left is priority  |
| Right     | All rows from right | Left side becomes `NaN`             | Yes              | ≥ right table | Right is priority |
| Outer     | All rows from both  | Both sides get `NaN` where no match | No               | Largest       | Keep everything   |
