### Pandas Series

A Pandas Series is like a column in a table. It is a 1-D array holding data of any type.

### 1. Series from lists

````python
# String

country = ['India', 'Pakistan', 'USA', 'Nepal', 'Srilanka']

pd.Series(country)    # # pd.Series is a class in pandas. Here we create a Series object by passing the list 'country' as data

# The list country is passed as data to the Series constructor, so Pandas converts the list into a 1-dimensional labeled array

# Object means a general Python value stored in memory (like string, list, etc.)
# In Pandas, strings are stored as Python string objects, so the column type becomes dtype: object
````

````python
# integers

runs = [16,33,77,99,21]
runs_series = pd.Series(runs)
print(runs_series)
````

````python
# custom index

mark = [97, 67, 60, 70]
subjects = ['maths','english','science','hindi']

pd.Series(marks, index=subjects)
````

````python
# Setting a name

marks = pd.Series(mark, index=subjects, name='vbp ke marks')
marks
````

#### Crunch

##### 1. `movies.index` is not a list

It is a **Pandas Index object**, not a Python list.

##### 2. The index already exists in the DataFrame

The index is already stored inside the DataFrame.\
Accessing `movies.index` does **not create a new list or copy the
data**.

##### 3. What happens in the loop

``` python
for i in movies.index:
    print(i)
```

Python creates an **iterator** and retrieves **one index value at a
time**.

Internally it behaves like this:

``` python
iterator = iter(movies.index)

while True:
    i = next(iterator)
    print(i)
```

##### 4. How iteration works

Index object\
↓\
Iterator created\
↓\
Next value fetched\
↓\
Printed\
↓\
Repeat

##### 5. RangeIndex memory optimization

If the DataFrame uses a **RangeIndex**, pandas does not store all
numbers.\
Instead it stores only:

    start
    stop
    step

and generates index values when needed.

##### 6. One-line idea

The index already exists in the DataFrame, and the loop simply
**iterates over it using an iterator without creating a new list**.
