### `Q-3` Write a Pandas program to compare the elements of the two Pandas Series.

### `Q-5.`Write a function to change the data type of given a column or a Series. Function takes series and data type as input, returns the converted series.
```
series = pd.Series([1,2,'Python', 2.0, True, 100])
change to float type data
```
Note: Read about pd.to_numeric()

````python
series = pd.Series([1,2,'Python', 2.0, True, 100])
print(series)
````

````python
def change_dtype(series, dtype):
    return pd.to_numeric(series, errors='coerce').astype(dtype)


series = pd.Series([1,2,'Python', 2.0, True, 100])
result = change_dtype(series, 'float32')

print(result)
````

### `Q-9`
    i. Read `items.csv` making `item_name` as index.
    ii. Show no of nan values
    ii. Item price is given in $, so convert it to rupees without currency symbol.
    iii. Make data type of newly made series as float.
    iv. Fill nan with mean of the series


How csv file looks

```
item_name	item_price
Chips and Fresh Tomato Salsa	$2.39
Izze	$3.39
Nantucket Nectar	$3.39
Chips and Tomatillo-Green Chili Salsa	$2.39
Chicken Bowl	$16.98

```

````python
# 1. Read `items.csv` making `item_name` as index

products = pd.read_csv('datasets/items.csv', index_col='item_name')
products
````

````python
# 2. Show no of nan values

products.isnull().sum()
````

````python
# 3. Item price is given in $, so convert it to rupees without currency symbol
````

````python
products["item_price"] = products["item_price"].str.replace('$','').astype(float)*83
````

````python
products
````

````python
#  4. Make data type of newly made series as float.

# already did it

# products['item_price'].astype(float)

products['item_price'].dtype
````

````python
# 5. Fill nan with mean of the series

products['item_price'] = products['item_price'].fillna(products['item_price'].mean())
products.isnull().sum()
````

fillna( ): replaces NaN values with the mean.

### `Q-10`:

````python
# 1. Find mean price
products['item_price'].mean()
````
