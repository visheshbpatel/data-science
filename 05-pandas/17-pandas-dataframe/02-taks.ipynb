````python
def team1_vs_team2(team1, team2):
    
    mask = ((ipl['Team1'] == team1) & (ipl['Team2'] == team2)) | ((ipl['Team1'] == team2) & (ipl['Team2'] == team1))
    
    temp = ipl[mask].copy()
    
    temp['LosingTeam'] = temp.apply(
        lambda x: x['Team2'] if x['WinningTeam'] == x['Team1'] else x['Team1'],
        axis=1
    )
    
    records = {
        "Team1": temp['Team1'],
        "Team2": temp['Team2'],
        "WinningTeam": temp['WinningTeam'],
        "LosingTeam": temp['LosingTeam']
    }
    
    result = pd.DataFrame(records)
    
    return result

team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')
````
