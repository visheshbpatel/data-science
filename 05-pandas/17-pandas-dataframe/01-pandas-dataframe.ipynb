````python
# Action movies with rating higher than 7.5

mask1 = movies['genres'].str.split('|').apply(lambda x:'Action' in x)

mask2 = movies['imdb_rating'] > 7.5

movies[mask1 & mask2]
````

> flow of str, split and apply

```python
movies['genres']        # selects the genres column
.str                    # allows applying string methods on the column
.split('|')             # splits genre strings into lists using '|'
.apply(lambda x: 'Action' in x)   # checks if 'Action' exists in each list
```

````python
mask1 = movies['genres'].str.contains('Action')

mask2 = movies['imdb_rating'] > 7.5

movies[mask1 & mask2]
````

````python
# Category is a dtype used to store data that has a limited number of unique values by converting them into integer codes.

# Actual values --> stored once (categories)

# Column -->  stores integers (codes)
````

````python
import pandas as pd

s = pd.Series(['Akshay','Rahul','pandey','Akshay'])

s = s.astype('category')

print(s)
print(s.dtype)
````

````python
# Internally:

# Categories: ['Akshay', 'Rahul' 'pandey']
# Codes:      [0, 1, 2, 0]
````

````python
# object dtype --> pointers --> slow  
# category dtype --> integers --> fast
````

````python
# Categories are stored in order of first appearance, not sorted by default.
````
