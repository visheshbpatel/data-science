### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
fifa = pd.read_csv('dataset/Fifa Worldcup 2022 - Sheet1.csv')
````

````python
fifa
````

````python
fifa.groupby(['Team']).groups
````

````python
import pandas as pd

def z_normalization(group, cols_to_perform):
    group = group.copy()  # avoid SettingWithCopyWarning
    
    for col in cols_to_perform:
        std = group[col].std()
        if std != 0:
            group[col] = (group[col] - group[col].mean()) / std
        else:
            group[col] = 0  # handle division by zero
    
    return group


cols = [
    'Passes',
    'Passes Completed',
    'Attempted Line Breaks',
    'Completed Line Breaks'
]


fifa_z = fifa.copy()

for col in cols:
    mean = fifa.groupby('Team')[col].transform('mean')
    std = fifa.groupby('Team')[col].transform('std')
    
    fifa_z[col] = (fifa[col] - mean) / std


fifa_z = fifa_z.reset_index(drop=True)


result = fifa_z.groupby('Team').agg({
    'Passes': ['min', 'max'],
    'Yellow Cards': ['min', 'max', 'mean'],
    'Attempted Line Breaks': ['min', 'max', 'std'],
    'Possession (%)': 'mean'
})

result.unstack()

result
````

### `Q-6` Most Successful bowler against any batsman. Find that pair of bowler and batsman.
> Most Successful in terms of dissmissal. A bowler who have dissmissed any batsman most no of times. If any two pairs have same no of dissmisal, consider runs conceded by bowler to that batsman. Those who have concede lesser runs is more successful.

````python
ipl_del.head()
````

````python
df = ipl_del.copy()

# filter wickets (exclude run-out)
wickets = df[
    (df['isWicketDelivery'] == 1) &
    (df['kind'] != 'run out')
]

# dismissals
dismissals = wickets.groupby(['bowler', 'player_out']).size().reset_index(name='wickets')

# runs conceded
runs = df.groupby(['bowler', 'batter'])['batsman_run'].sum().reset_index(name='runs')

# merge
result = dismissals.merge(
    runs,
    left_on=['bowler', 'player_out'],
    right_on=['bowler', 'batter']
)

# clean
result = result.rename(columns={'player_out': 'batsman'})
result = result.drop(columns=['batter'])

# sort
result = result.sort_values(by=['wickets', 'runs'], ascending=[False, True])

# answer
result.head()
````

### `Q-7`: Most successful batting pair in IPL. Batting pair who have scored most runs playing together.


````python
ipl_del.head()
````

````python
pairs = ipl_del.copy()


pairs['pair'] = pairs.apply(
    lambda x: tuple(sorted([x['batter'],x['non-striker']])), axis=1
)


partnership = pairs.groupby('pair').agg(
    runs = ('batsman_run', 'sum'),
    extras = ('extras_run', 'sum')
)

partnership['total_runs'] = partnership['runs'] + partnership['extras']


partnership.sort_values(by='total_runs', ascending=False).head()
````

### `Q-8:` Make a dataframe for all batting pairs played together.
```
Batsman1 Batsman2 Runs Avg StrikeRate
```

> Just to ease this question you can count wide-balls for strike rate.

````python
ipl_del.head(3)
````

````python
batting_pairs = ipl_del.copy()

batting_pairs['pairs'] = batting_pairs.apply(
    lambda x: tuple(sorted([x['batter'], x['non-striker']])), axis=1
)

partnership = batting_pairs.groupby('pairs').agg(
    Batsman1 = ('pairs', lambda x: x.iloc[0][0]),
    Batsman2 = ('pairs', lambda x: x.iloc[0][1]),
    Runs = ('batsman_run','sum'),
    Balls = ('batsman_run','count')
    ).reset_index()

partnership['Avg'] = (partnership['Runs'] / partnership['Balls']).round(2)

partnership['StrikeRate'] = ((partnership['Runs'] / partnership['Balls']) * 100).astype(int)


partnership.drop(columns = ['pairs','Balls'], inplace=True)


partnership.sort_values(by='Runs', ascending=False)

````

````python
batting_pairs = ipl_del.copy()

# Create pairs (vectorized way)
batting_pairs['pairs'] = list(
    map(lambda x, y: tuple(sorted((x, y))),
        batting_pairs['batter'],
        batting_pairs['non-striker'])
)


# Grouping
partnership = batting_pairs.groupby('pairs').agg(
    Runs=('batsman_run', 'sum'),
    Balls=('batsman_run', 'count')
).reset_index()


# Split tuple into two columns
partnership[['Batsman1', 'Batsman2']] = pd.DataFrame(
    partnership['pairs'].tolist(), index=partnership.index
)


# Metrics
partnership['Avg'] = (partnership['Runs'] / partnership['Balls']).round(2)
partnership['StrikeRate'] = ((partnership['Runs'] / partnership['Balls']) * 100).round(2)


# Final cleanup
partnership = partnership.drop(columns=['pairs', 'Balls'])


# Sort
partnership = partnership.sort_values(by='Runs', ascending=False)


partnership
````
