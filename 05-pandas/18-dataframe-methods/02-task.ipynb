````python
# Replace old team names
ipl.replace("Delhi Daredevils","Delhi Capitals", inplace=True)
ipl.replace("Kings XI Punjab","Punjab Kings", inplace=True)
ipl.replace("Rising Pune Supergiants","Rising Pune Supergiant", inplace=True)

# Remove matches with no result
ipl = ipl[ipl['WinningTeam'].notnull()]

# Matches played by each team
real_ipl = pd.concat([ipl['Team1'], ipl['Team2']]).value_counts()

real_ipl = real_ipl.reset_index()
real_ipl.columns = ["TeamName","MatchesPlayed"]


# Win %
def winningPer(row):
    team = row['TeamName']
    total = row['MatchesPlayed']

    wins = len(ipl[ipl['WinningTeam'] == team])

    return round((wins/total)*100,2)

real_ipl['Win%'] = real_ipl.apply(winningPer, axis=1)


# Home Win %
def winningPerHome(row):
    team = row['TeamName']

    home_matches = len(ipl[ipl['Team1'] == team])
    home_wins = len(ipl[(ipl['Team1'] == team) & (ipl['WinningTeam'] == team)])

    return round((home_wins/home_matches)*100,2)

real_ipl['HomeWin%'] = real_ipl.apply(winningPerHome, axis=1)


# Away Win %
def winningPerAway(row):
    team = row['TeamName']

    away_matches = len(ipl[ipl['Team2'] == team])
    away_wins = len(ipl[(ipl['Team2'] == team) & (ipl['WinningTeam'] == team)])

    return round((away_wins/away_matches)*100,2)

real_ipl['AwayWin%'] = real_ipl.apply(winningPerAway, axis=1)


# Sort by Win %
real_ipl = real_ipl.sort_values(by='Win%', ascending=False)


real_ipl
````

````python
def point_table(season):
    temp = ipl[ipl["Season"] == season]

    # 1. Matches Played
    matches_played = pd.concat([temp['Team1'], temp['Team2']]).value_counts()

    
    df = matches_played.reset_index()
    df.columns = ['TeamName', 'MatchesPlayed']
    df.set_index('TeamName', inplace=True)

    
    # 2. Matches Won
    matches_won = temp['WinningTeam'].value_counts()
    df['MatchesWon'] = matches_won

    
    # 3. No Result (winner is NaN)
    no_result_matches = temp[temp['WinningTeam'].isna()]

    no_result = pd.concat([no_result_matches['Team1'], no_result_matches['Team2']]).value_counts()
    df['NoResult'] = no_result

    
    # Fill NaN with 0
    df.fillna(0, inplace=True)

    
    # Convert to int
    df = df.astype(int)

    
    # 4. Points
    df['Points'] = df['MatchesWon'] * 2 + df['NoResult'] * 1

    
    # 5. Sorting
    df.sort_values(by=['Points', 'TeamName'], ascending=[False, True], inplace=True)

    return df


table = point_table('2022')
table
````

````python
def point_table(season):
    temp = ipl[ipl["Season"] == "2022"].copy()
    temp['LossingTeam'] = temp.apply(lambda x: x['Team1'] if x['Team1'] != x['WinningTeam'] else x['Team2'], axis=1)

    
    matches_played = pd.concat([temp['Team1'], temp['Team2']]).value_counts()

    
    df = matches_played.reset_index()
    df.columns = ['TeamName', 'MatchesPlayed']
    df.set_index('TeamName', inplace=True)

    matches_won = temp['WinningTeam'].value_counts()
    df['MatchesWon'] = matches_won

    no_result_matches = temp[temp['WinningTeam'].isna()]

    no_result = pd.concat([no_result_matches['Team1'], no_result_matches['Team2']]).value_counts()
    df['NoResult'] = no_result

    df.fillna(0, inplace=True)

    
    df = df.astype(int)

    
    df['Points'] = df['MatchesWon'] * 2 + df['NoResult'] * 1

    df.sort_values(by=['Points', 'TeamName'], ascending=[False, True], inplace=True)

    df['SeasonPosition'] = df['Points'].rank(method='first', ascending=False).astype(int)

    df['SeasonPosition'] = df['SeasonPosition'].astype(str)


    # # final
    final = temp[temp['MatchNumber'] == 'Final']['WinningTeam']
    df.loc[final[0], 'SeasonPosition'] = 'Final'


    # Runner
    runner = temp[temp['MatchNumber'] == 'Final']['LossingTeam']
    df.loc[runner[0], 'SeasonPosition'] = 'Runner'


    #Qualifier2
    qualifier2 = temp[temp['MatchNumber'] == 'Qualifier 2']['LossingTeam']
    df.loc[qualifier2, 'SeasonPosition'] = 3

    # Eliminator
    eliminator = temp[temp['MatchNumber'] == 'Eliminator']['LossingTeam']
    df.loc[eliminator, 'SeasonPosition'] = 4


    

    return df


table = point_table('2022')
table
````

````python
def point_table(season):
    temp = ipl[ipl["Season"] == season].copy()

    # Losing team (handle NaN properly)
    temp['LosingTeam'] = temp.apply(
        lambda x: None if pd.isna(x['WinningTeam'])
        else (x['Team1'] if x['Team1'] != x['WinningTeam'] else x['Team2']),
        axis=1
    )

    
    # Matches Played
    matches_played = pd.concat([temp['Team1'], temp['Team2']]).value_counts()
    df = matches_played.reset_index()
    df.columns = ['TeamName', 'MatchesPlayed']
    df.set_index('TeamName', inplace=True)

    
    # Matches Won
    df['MatchesWon'] = temp['WinningTeam'].value_counts()

    
    # No Result
    no_result_matches = temp[temp['WinningTeam'].isna()]
    no_result = pd.concat([no_result_matches['Team1'], no_result_matches['Team2']]).value_counts()
    df['NoResult'] = no_result

    df.fillna(0, inplace=True)
    df = df.astype(int)

    
    # Points
    df['Points'] = df['MatchesWon'] * 2 + df['NoResult']

    
    # Sort
    df.sort_values(by=['Points' ,'TeamName'], ascending=[False, True], inplace=True)
    # df.sort_values(by=['Points','MatchesPlayed', 'TeamName'], ascending=[False, True, True], inplace=True)


    # Rank for others
    df['SeasonPosition'] = df['Points'].rank(method='first', ascending=False).astype(int)

    
    # avoid warnings for stings(int brefore code)
    df['SeasonPosition'] = df['SeasonPosition'].astype(object)


    
    # Winner
    winner = temp[temp['MatchNumber'] == 'Final']['WinningTeam'].iloc[0]
    df.at[winner, 'SeasonPosition'] = 'Winner'

    
    # Runner
    runner = temp[temp['MatchNumber'] == 'Final']['LosingTeam'].iloc[0]
    df.at[runner, 'SeasonPosition'] = 'Runner'

    
    # Qualifier 2 loser - 3rd
    q2_loser = temp[temp['MatchNumber'] == 'Qualifier 2']['LosingTeam'].iloc[0]
    df.at[q2_loser, 'SeasonPosition'] = 3

    
    # Eliminator loser - 4th
    elim_loser = temp[temp['MatchNumber'] == 'Eliminator']['LosingTeam'].iloc[0]
    df.at[elim_loser, 'SeasonPosition'] = 4

    return df

point_table('2022')
````
