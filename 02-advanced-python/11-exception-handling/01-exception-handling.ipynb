***There are 2 stages where error may happen in a program***
- *During compilation -> Syntax Error*
- *During Execution -> Exception*

## 1. Syntax Error

- Something in the program is not written according to the program grammar.
- Error is raised by the interpreter/complier
- You can solve it by rectifying the program

```python
print 'hello world'
#---> SyntaxError: Missing parentheses in call to 'print'. Did you mean print(...)?
```

Other Examples of syntax error

- Leaving symbols like colon, brackets
- Misspelling a keyword
- Incorrect indentation
- empty if/else/loops/class/function

```python
class Bank:

    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount<0:
            raise ValueError("Amount can not be negative")         # Exception is a class
        if self.balance<amount:
            raise ValueError("Paise nahi hai tere pass")
        self.balance-=amount

obj = Bank(10000)


try:
    obj.withdraw(5000)
except Exception as e:
    print(e)
else:
    print("Remaining Balance:", obj.balance)

try:
    obj.withdraw(15000)
except Exception as e:
    print(e)
else:
    print("Remaining Balance:", obj.balance)


try:
    obj.withdraw(-1000)
except Exception as e:
    print(e)
else:
    print("Remaining Balance:", obj.balance)


# Exception is a class , so when a error is raised then it creates a object of it which is stored by the e (Exception as e ),
# and the in except we print that e
# Basically raise throws an error to except , and the except catch's that error and stores it into e
# e now holds the exception object
```

```python
# custom error Exception class
class MyException(Exception):

    def __init__(self,message):
        print(message)


class Bank:

    def __init__(self, balance):
        self.balance = balance

    def withdraw(self, amount):
        if amount<0:
            raise MyException("Amount can not be negative")
        if self.balance<amount:
            raise MyException("Paise nahi hai tere pass")
        self.balance-=amount

obj = Bank(10000)


try:
    obj.withdraw(15000)
except MyException as e:
    pass
else:
    print("Remaining Balance:", obj.balance)



try:
    obj.withdraw(-5000)
except MyException as e:
    pass
else:
    print("Remaining Balance:", obj.balance)



try:
    obj.withdraw(5000)
except MyException as e:
    pass
else:
    print("Remaining Balance:", obj.balance)
```
