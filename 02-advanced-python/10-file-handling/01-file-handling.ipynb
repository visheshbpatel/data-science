## Some Theory

Type of Data used for I/O:
- Text - '12345' as a sequence of unicode chars
- Binary - 12345 as a sequence of bytes of its binary equivalent

Hence there are 2 file types to deal with
- Text files - All program files are text files
- Binary Files - Images, music, video, exe files

#### How Files I/O is done in most programming languages
- open a file
- Read/Write data
- Close the file

## Writing a file

```python
# case-1 if the file is not present
f = open('sample.txt', 'w')   # here we have created that file
f.write("Namaste Duniya!!!")
f.close()

# f.write('Hello')     #---> ValueError: I/O operation on closed file.
# since file is closed hence this will not work
```

```python
# Case-2 if the file is already present

f = open('sample.txt', 'w')
f.write('Nitish Sir is GOAT!!!')
f.close()
```

***How exactly open() works?***

1. open() does:

2. Ask OS to open file

3. OS returns file descriptor

4. Python wraps it into file object

5. That object is stored in variable f

6. You use its methods to read/write

```python
# Introducing append mode

f = open('sample.txt', 'a')  # a for append
f.write('\n#CampusX')
f.close()
```

```python
# Write Lines

L = ['hello\n', 'hii\n', 'how are you\n', 'I am fine\n']

f = open('sample1.txt', 'w')
f.writelines(L)
f.close()
```

> We use close() to:

- Flush buffer to disk

- Release OS file descriptor

- Prevent file locking issues

- Avoid memory/resource leaks

- Follow clean programming practice

```python
# Reading from files
# --> using read()

f = open('sample.txt','r')
s = f.read()
print(s)
f.close()
```

```python
# Reading upto n chars

f = open('sample.txt', 'r')
s = f.read(10)
print(s)
f.close()
```

```python
# readline() --> to read line by line

f = open('sample.txt', 'r')
print(f.readline(), end='')
print(f.readline(), end='')
f.close()
```

```python
# Reading entire using readline

f = open('sample.txt', 'r')

while True:

    data = f.readline()

    if data =="":
        break
    else:
        print(data, end='')

f.close()
```

## Using Context Manager (With)

- It's a good idea to close a file after usage as it will free up the resources
- If we don't close it, garbage collector would close it
- with keyword closes the file as soon as the usage is over

```python
# with
with open('sample1.txt', 'w') as f:
    f.write('Me hu patel')
```

```python
# f.write('Hello')    #---> ValueError: I/O operation on closed file.
# The file is closed by using 'with' , no need to close it
```

```python
# try f.read() now

with open('sample.txt', 'r') as f:
    print(f.read())
```

```python
# moving within a file --> 10 char then 10 char

with open('sample.txt', 'r') as f:
    print(f.read(10))
    print(f.read(10))
    print(f.read(10))


# read(n) reads n characters
# After reading, pointer moves forward
# Next read continues from new position
```

```python
# benefits ? ---> to load a big file in memory

big_L = ['Hello world' for i in range(1000)]

with open('big.txt', 'w') as f:
    f.writelines(big_L)

```

```python
with open('big.txt', 'r') as f:

    chunk_size = 11

    chunk = f.read(chunk_size)

    while len(chunk) > 0:
        print(chunk, end="  -- ")
        chunk = f.read(chunk_size)
```

```python
# seek and tell function

with open('big.txt', 'r') as f:
    print(f.read(11))
    
    print(f.tell())       # tell() returns:Current file pointer position
    f.seek(0)             # seek(pos) → moves pointer

    print(f.read(11))
    print(f.tell())
```

```python
# seek during write

with open('sample1.txt', 'w') as f:
    f.write('Hello')
    f.seek(0)
    f.write('X')
```

## Serialization and Deserialization

- ***Serialization*** -  Process of converting python data types to JSON format
- ***Deserialization*** - Process fo converting JSON to python data types

<img src="https://pageflows.com/resources/wp-content/uploads/2024/06/JSONExample.png">

```python
# Serialization using json module
# List

import json

L = [1,2,3,4]

with open('demo.json', 'w') as f:
    json.dump(L,f)   # json.dump() converts it into JSON format
```

```python

d = {
    'name':'vbp',
    'age':21,
    'gender':'male',
}


with open('demo2.json', 'w') as f:
    json.dump(d,f, indent=4)
```

```python
# Deserialization

import json

with open('demo.json', 'r') as f:
    d = json.load(f)     # json.load() converts JSON to Python object.
    print(d)
    print(type(d))
```

```python
# Serialize and deserialize tuple

import json

t = (1,2,3,4,5)

with open('demo3.json','w') as f:
    json.dump(t,f)
```

## Serializing and Deserializing custom objects

```python
class Person:

    def __init__(self, fname, lname, age, gender):
        self.fname = fname
        self.lname = lname
        self.age = age
        self.gender = gender

        

# format to print in
# --> vishesh patel age-> 21 gender-> male
```

```python
person = Person('visheh','patel','21','male')
```

```python
# As a string

import json

with open("demo5.json", 'w') as f:
    pass
    # json.dump(person, f)    #---->  TypeError: Object of type Person is not JSON serializable
```

```python
# correct way to do it

import json

def show_object(person):
    if isinstance(person,Person):
        return "{} {} age-> {} gender-> {} ".format(person.fname, person.lname, person.age, person.gender)
with open("demo5.json", 'w') as f:
    json.dump(person, f, default=show_object)
```

```python
# same thing in dict formate

import json

def show_object(person):
    if isinstance(person,Person):
        return {'name':person.fname + ' ' + person.lname, 'age':person.age, 'gender':person.gender}
with open("demo6.json", 'w') as f:
    json.dump(person, f, default=show_object, indent=4)
```

```python
# Deserializing 

import json

with open('demo6.json','r') as f:
    d = json.load(f)
    print(d)
    print(type(d))
```

## Pickling

***The process of converting a Python object into a binary byte stream so that it can be stored or transmitted and reconstructed later.***

***Unpickling is the process of converting a pickled (binary) byte stream back into a Python object.***

```python
class Person:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print("Hii my name is", self.name, "and I am",self.age,"year old")
```

```python
p = Person('vbp',21)

#converting into binary

import pickle
with open('person.pkl', 'wb') as f:
    pickle.dump(p,f)
```

```python
# pickle dump
#converting into binary

import pickle
with open('person.pkl', 'wb') as f:
    pickle.dump(p,f)
```

```python
# pickle load

import pickle
with open('Person.pkl', 'rb') as f:
    p_obj = pickle.load(f)

p_obj.display_info()
```

## Pickle Vs JSON

Both Pickle and JSON are used for serialization and deserialization.

- ***Pickle lets the user to store data in binary format. JSON lets the user store data in a human-readable text format.***

JSON is a text-based, language-independent serialization format mainly used for data exchange, while Pickle is a Python-specific binary serialization format capable of storing complex Python objects but unsafe for untrusted sources.
    

- JSON = Universal language

- Pickle = Python memory freezer
