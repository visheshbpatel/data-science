## 1. Namespaces & Scopes

### 1.1 Namespaces

A namespace is a space that holds names(identifiers). Programmatically speaking. Namespace are dictionary of identifiers(keys) and their object(values)

There are 4 types of namespaces:
- Builtin Namespace
- Global Namespaces
- Enclosing Namespace
- Local Namespace


namespace:
```python
a=2
b=3
{'a': 2, 'b': 3}
```

A namespace is a container that maps names (identifiers) to objects.

Technically:

A namespace is implemented as a dictionary where
-  Key --> identifier name
-  Value --> object

###  1.2 Scope and LEGB Rule

A scope is a textual region of python program where a namespace is directly accessible.


*Local Enclosing Global Builtin(LEGB):*

The interpreter searches for a name from the inside out, looking in the local, encloding, global, and finally the built-in scope. If the interpreter doesn't find the name in any of these locations, then Python raises a NameError exception.

Local Enclosing Global Builtin(LEGB)


Namespace --> where names are stored

Scope --> where names can be accessed

````python
# local and global --> local does not have but global have

a = 2


def temp():
    print(a)

temp()
print(a)


# If the Python interpreter does not find a variable in the local scope, it searches in the enclosing scope, then in the global scope, and finally in the built-in scope.
# If it is not found in any of these scopes, Python raises a NameError.
````

````python
# local and global  --> editing global

a = 2


def temp():
    # a+=1           #---> UnboundLocalError: cannot access local variable 'a' where it is not associated with a value
    print(a)

temp()
print(a)


# we can read global varibale but can't write it
````

````python
# using keyword global we can access global variable and can make changes

a = 2

def temp():
    global a        #--> without global it will give unboundLocalError
    a+=1           
    print(a)

temp()
print(a)


# we should not do it , it is not recommended 
# using too much global is not a good programming practice
````

````python
# local and global --> global created inside local

def temp():
    global a       # a was not there in global scope , but because of this a is created in global scope
    a = 1          # we have created global variable a from local scope
    print(a)

temp()
print(a)


# global does NOT Move variable from local to global

# It DOES Tell Python to bind that name in the global namespace
````

````python
# local and global --> function parameter is local

def temp(z):        # z is a local variable, program/functions parameters are local variable
    print(z)


a=5                 # a is global varibale
temp(a)
print(a)
````

````python
def temp():
    a=3
    # global a        #---> SyntaxError: name 'a' is assigned to before global declaration
    a = 1          
    print(a)

temp()
print(a)


# If we use global, it must appear before any use or assignment of that variable
# we cannot first create a local variable and then declare it global
````

````python
# renaming built-ins

L = [1,2,3]      

# max(L)                          #--> TypeError: max() takes 0 positional arguments but 1 was given
 
def max():                        # overriding
    print("Hello")

# max(L)                          #--> TypeError: max() takes 0 positional arguments but 1 was given
````

1. Local --> not found

## 2. Decorators

A decorator is a function that:

1. Takes another function as input

2. Adds extra functionality(decoration) to it

3. Returns the modified function

4. Without permanently changing the original function’s code.

A decorator is a function that receives another function as an argument, extends its behavior, and returns a new function.

***This can happen only beacause python functions are 1st class citizens***

There are 2 types of decorators available in python

- built in decorators like `@staticmethod`, `@classmethod`, `@abstractmethod` and `@property` etc
- User defined decorators that we programmers can create according to our needs

````python
# python functions are 1st class citizens

def P_func():
    print("Hello")

P_func      # reference to function
P_func()    # calling the function

a = P_func()
a
````

````python
del P_func
# p_func()      #---> NameError: name 'func' is not defined
````

````python
def modify(func,num):
    return func(num)

def square(num):
    return (num**2)

modify(square,2)
````

````python
# simple exaxmple

def my_decorator(func):
    def wrapper():
        print("*************************")
        func()
        print("*************************")
    return wrapper

def hello():
    print("Hello")

def display():
    print("Hello vbp")
    

a = my_decorator(hello)
a()

b = my_decorator(display)
b()
````

````python
def outer():
    a=5
    def inner():
        print(a)
    return inner

b = outer()
b()
````

*outer( ) is called and inside it:*
- a = 5
- inner( ) defined 
- returns inner

*Even though outer() has finished execution,inner still remembers a = 5.*



Because Python forms a closure.

A closure is:

*A function that remembers variables from its enclosing scope even after the outer function has finished execution.*

In our case:

- inner() remembers a

- The execution frame of outer() is removed, but the variables needed by inner() are preserved inside a closure object.


Normally:

Local variables disappear after function ends.

But here:

- Since inner is returned

- And inner uses a

- Python keeps a alive

- This is called a free variable stored in a closure.


    
*Python stores a inside:*

```python
b.__closure__
```

That closure cell contains the value 5.

So a is not kept in the normal stack frame anymore
it is moved into a closure cell.

````python
print(b.__closure__)
````

````python
# Proof that b exists in closure

print(b.__closure__[0].cell_contents)
````

````python
# Better syntax

def my_decorator(func):
    def wrapper():
        print("*************************")
        func()
        print("*************************")
    return wrapper

@my_decorator
def hello():
    print("Hello")

    
hello()
````

Step 2:

````python
import time

def timer(func):
    def wrapper(*args):
        start = time.time()
        func(*args)
        print(f"Time take by {func.__name__} {time.time()-start} secs")
    return wrapper


@timer
def square(num):
    time.sleep(1)
    print(num**2)


@timer
def power(a,b):
    print(a**b)


square(2)
power(5,2)
````

````python
def sanity_check(data_type):
    def outer_wrapper(func):
        def inner_wrapper(*args):
            if isinstance(args[0], data_type):   # better practice
                return func(*args)               # return result
            else:
                raise TypeError("This datatype is not supported")
        return inner_wrapper
    return outer_wrapper


@sanity_check(int)            #--> internally: square = sanity_check(int)(square)
def square(num):
    return num**2

print(square(2))
# square('5')         #--> TypeError: This datatype is not supported
````

````python
def sanity_check(data_type):
    def outer_wrapper(func):
        def inner_wrapper(*args):
            if isinstance(*args, data_type):        #--> this can also work , using *args and avoding args[0] but only works for a single value 
                                                    # if there are 2 valeus , like: (2,5, int) it won't work, so need to be more carefull
                return func(*args)               
            else:
                raise TypeError("This datatype is not supported")
        return inner_wrapper
    return outer_wrapper


@sanity_check(int)            
def square(num):
    return num**2

print(square(2))
````

````python
@sanity_check(str)
def greet(name):
    print(f"Hello {name}")


greet('Vishesh')
# greet(101)           #--> TypeError: This datatype is not supported
````

### Parameterized Decorator -- Step-by-Step Flow Explanation

#### Step-by-Step Flow

##### 1. Python Reads the Decorator Definition

Nothing executes yet.\
Python just stores `sanity_check` in memory.

------------------------------------------------------------------------

##### 2. Python Reaches the Decorator Line

``` python
@sanity_check(int)
def square(num):
```

This is internally converted to:

``` python
def square(num):
    return num**2

square = sanity_check(int)(square)
```

Now execution begins.

------------------------------------------------------------------------

##### 3. First Call Happens

``` python
sanity_check(int)
```

Execution enters:

``` python
def sanity_check(data_type):
```

Here:

    data_type = int

It defines `outer_wrapper` and returns it.

So:

    sanity_check(int) → returns outer_wrapper

------------------------------------------------------------------------

##### 4. Second Call Happens

Now this runs:

``` python
outer_wrapper(square)
```

Because:

    square = sanity_check(int)(square)

Execution enters:

``` python
def outer_wrapper(func):
```

Here:

    func = original square function

Inside it: - `inner_wrapper` is created - and returned

So now:

    square = inner_wrapper

The original `square` function is replaced by `inner_wrapper`.

But the original function still exists inside the closure as `func`.

------------------------------------------------------------------------

#### Current Memory State

    square → inner_wrapper
                 ↓
               uses:
                 func → original square
                 data_type → int

`inner_wrapper` remembers: - `func` - `data_type`

This is called a closure.

------------------------------------------------------------------------

##### 5. Now This Line Runs

``` python
print(square(2))
```

But remember:

    square = inner_wrapper

So this actually runs:

``` python
inner_wrapper(2)
```

------------------------------------------------------------------------

##### 6. Execution Inside inner_wrapper

``` python
if isinstance(args[0], data_type):
```

Here:

    args = (2,)
    data_type = int

So:

    isinstance(2, int) → True

Then:

``` python
return func(*args)
```

Which means:

``` python
return original_square(2)
```

------------------------------------------------------------------------

##### 7. Original Function Runs

``` python
return num ** 2
```

So:

    2 ** 2 → 4

That 4 returns to: - `inner_wrapper` - then to `print()`

------------------------------------------------------------------------

#### Final Output

    4

------------------------------------------------------------------------

##### What If You Call

``` python
square('5')
```

Flow:

    inner_wrapper('5')

Check:

    isinstance('5', int) → False

So:

    raise TypeError

------------------------------------------------------------------------

#### Visual Execution Diagram

    @sanity_check(int)
            ↓
    sanity_check(int)
            ↓
    returns outer_wrapper
            ↓
    outer_wrapper(square)
            ↓
    returns inner_wrapper
            ↓
    square = inner_wrapper
            ↓
    square(2)
            ↓
    inner_wrapper(2)
            ↓
    original square(2)
            ↓
    returns 4

------------------------------------------------------------------------

#### Why This Works

This works because:

1.  Functions are first-class objects
2.  Functions can return functions
3.  Inner functions remember outer variables (closure)

Even after `sanity_check` finishes,
`data_type` still exists inside `inner_wrapper` because of closure.


### Decorator Execution Timing -- Explanation

Even if you never call:

``` python
print(square(2))
```

this line is still executed:

``` python
square = sanity_check(int)(square)
```

------------------------------------------------------------------------

#### When Does the Decorator Execute?

Decorators execute at function definition time, not at function call
time.

So this reassignment happens while the file is being executed from top
to bottom, even before you ever call `square()`.

------------------------------------------------------------------------

#### What Happens Even Without Calling square(2)?

The following still happens:

1.  `sanity_check(int)` runs
2.  Returns `outer_wrapper`
3.  `outer_wrapper(square)` runs
4.  Returns `inner_wrapper
5.  `square` now points to `inner_wrapper`

So the decoration is already done.

------------------------------------------------------------------------

#### What Does NOT Happen?

The following do NOT happen until you call `square()`:

-   `inner_wrapper` execution
-   Type checking
-   The actual squaring

Because those occur only at function call time.

------------------------------------------------------------------------

#### Timeline of Execution

When Python runs your file:

1.  Define `sanity_check`
2.  Reach `@sanity_check(int)`
3.  Execute decoration immediately
4.  Replace `square` with `inner_wrapper`
5.  Move to the next line

Calling `square()` later is a completely separate step.

````python
import time

def timer(func):
    def wrapper(*args):
        start = time.time()
        func(*args)
        print(f"Time take by {func.__name__} {time.time()-start} secs")
    return wrapper


@timer
def power(a,b):
    print(a**b)


power(5,2)
````

*args collects arguments into a tuple.

func(*args) unpacks them back.

The original function still receives arguments normally.

So print(a**b) works because the arguments are properly unpacked and passed to the original function.


----

Why Did It Unpack?

Because we wrote:

func(*args)

The * in a function call means:

"Take this tuple and spread its elements as separate positional arguments."
