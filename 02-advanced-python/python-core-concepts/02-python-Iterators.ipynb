## 1.What is an Iteration

Iteration is a general term for taking each item of something, one after another. Any time we use a loop, explicit or implicit, to go over a group of items, that is iteration.

````python
# Example
num = [1,2,3]

for i in num:
    print(i)
````

## 2. What is Iterator

An Iterator is an object that allows the programmer to traverse through a sequence of data without having to store the entire data in the memory

````python
# Example
L = [x for x in range(1,10000)]

#for i in L:
    #print(i*2)
    
import sys

print(sys.getsizeof(L)/64)

x = range(1,10000000000)

#for i in x:
    #print(i*2)
    
print(sys.getsizeof(x)/64)

````

## 6. Making our own for loop

````python
def mera_khudka_for_loop(iterable):
    
    iterator = iter(iterable)
    
    while True:
        try:
            print(next(iterator))
        except StopIteration:
            break           
````

````python
a = [1,2,3]
b = range(1,11)
c = (1,2,3)
d = {1,2,3}
e = {0:1,1:1}

mera_khudka_for_loop(a)
mera_khudka_for_loop(e)

````

## 8. Let's create our own range() function

````python
class mera_range:
    
    def __init__(self,start,end):
        self.start = start
        self.end = end
        
    def __iter__(self):
        return mera_range_iterator(self)
````

````python
class mera_range_iterator:
    
    def __init__(self,iterable_obj):
        self.iterable = iterable_obj
    
    def __iter__(self):
        return self
    
    def __next__(self):
        
        if self.iterable.start >= self.iterable.end:
            raise StopIteration
            
        current = self.iterable.start
        self.iterable.start+=1
        return current
````

### 9.1 Iteration, Iterator and Iterable

````python
# Iterable
numbers = [10, 20, 30, 40]

print("Iterable object:", numbers)

# Creating an iterator from iterable
iterator_obj = iter(numbers)

print("\nIterator created from iterable")

# Iteration using next()
print("\nIteration using next():")
print(next(iterator_obj))
print(next(iterator_obj))
print(next(iterator_obj))
print(next(iterator_obj))

# Uncommenting below will give StopIteration error
# print(next(iterator_obj))

# Iteration using for loop
print("\nIteration using for loop:")

for num in numbers:
    print(num)
````

#### Understanding Iterable, Iterator, and Iteration in Python


##### What is an Iterable

An **iterable** is an object that can be looped over.

Example

```python
numbers = [10, 20, 30, 40]
```

This list is an iterable because Python can access its elements one by one.

Examples of iterables

1. List  
2. Tuple  
3. String  
4. Set  
5. Dictionary  
6. Numpy arrays  

---

##### What is an Iterator

An **iterator** is an object that keeps track of the current position during iteration.

We create an iterator using the `iter()` function.

Example

```python
iterator_obj = iter(numbers)
```

Now `iterator_obj` is an iterator.

---

##### What is Iteration

**Iteration** is the process of accessing elements of an iterable one by one.

Example

```python
next(iterator_obj)
```

Each call to `next()` returns the next element from the iterator.

```
10
20
30
40
```

After the last element, Python raises a `StopIteration` error.

---

##### What Actually Happens Inside a For Loop

When you write

```python
for num in numbers:
    print(num)
```

Python internally does something similar to this

```python
iterator_obj = iter(numbers)

while True:
    try:
        num = next(iterator_obj)
        print(num)
    except StopIteration:
        break
```

So the **for loop automatically creates an iterator and keeps calling `next()`** until `StopIteration` occurs.

---

##### Short Summary

1. **Iterable**  
   An object that can be looped over.

2. **Iterator**  
   An object created from an iterable using `iter()`.

3. **Iteration**  
   The process of accessing elements one by one using `next()`.



````python
##### Program to Check Iterable and Iterator


from collections.abc import Iterable, Iterator

L = [1, 2, 3]

print("Is Iterable:", isinstance(L, Iterable))

it = iter(L)

print("Is Iterator:", isinstance(it, Iterator))
````

### 9.2 `iter()` Function in Python



`iter()` is a **built-in Python function** used to create an **iterator from an iterable object**.

In simple words, it converts something that can be looped over (like a list or string) into an **iterator that can return elements one by one**.

##### Syntax

```python
iter(object)
```

`object` must be an **iterable**.

---

##### Example

```python
numbers = [10, 20, 30, 40]

it = iter(numbers)

print(next(it))
print(next(it))
print(next(it))
print(next(it))
```

##### Output

```
10
20
30
40
```

Here:

1. `numbers` is an **iterable**
2. `iter(numbers)` creates an **iterator**
3. `next(it)` returns elements **one by one**

---
    
##### Step by Step Understanding

```python
numbers = [10,20,30]
```

This is an **iterable**.

Now we create an iterator.

```python
it = iter(numbers)
```

Now `it` is an **iterator object**.

When we call

```python
next(it)
```

Python returns the **next element**.

```
10
```

Next call

```
20
```

Next call

```
30
```

After the last element Python raises

```
StopIteration
```
---
    
##### Important Thing

A **for loop automatically uses `iter()` internally**.

When we write

```python
for num in numbers:
    print(num)
```

Python internally does something like this.

```python
it = iter(numbers)

while True:
    try:
        num = next(it)
        print(num)
    except StopIteration:
        break
```

##### Simple Summary

1. `iter()` converts an **iterable into an iterator**
2. The iterator returns values **one by one**
3. `next()` is used to get the **next value**
4. `for` loops use `iter()` automatically
